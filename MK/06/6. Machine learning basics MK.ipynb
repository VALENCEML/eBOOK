## 6.1. Што е машинско учење?

Машинското учење е подобласт на компјутерската наука која вклучува создавање алгоритми кои учат од примери за решавање на практични проблеми. Тоа вклучува креирање на податочно множество и градење статистички модел од него што ќе се користи за решавање на проблемот.

Еден пример за практичен проблем што може да се реши со користење на машинско учење е класификацијата на слики.
<ol>
  <li>Се собира множество на слики, секоја означена со соодветната класа на објекти (на пр. куче, мачка, автомобил итн.).</li>
<li>Се користи алгоритам за обука на модел на ова податочно множество, кој учи да ги препознава карактеристиките и шемите што разликуваат една класа на објекти од друга.</li>
<li>Откако ќе се обучи, моделот може да се користи за класификација на нови, претходно невидени слики во нивните правилни класи на објекти.</li>
</ol>

Во овој случај, моделот е статистички модел кој е изграден со користење на податочното множество на означени слики и се користи за класификација на нови слики врз основа на научените карактеристики и обрасци.

 ## 6.2. Надгледувано учење
Надгледуваното учење е тип на машинско учење каде што имаме податочно множество што ги вклучува и влезните податоци и соодветниот правилен излез. Целта е да се обучи модел на оваа база на податоци, за да може да прави предвидувања за нови, невидени податоци.

Замислете дека имате множество од слики од животни и секоја слика е означена (етикетирана) со името на животното на сликата (на пр. „куче“, „мачка“, „коњ“). Можеме да го  користиме ова множество за да обучиме модел на кој потоа ќе му дадеме нова слика (која не припаѓа во оригиналното множество) како влез и тој ќе предвиди кое животно е на сликата. Ова е пример за надгледувано учење.

Надгледуваното учење може да се користи за различни задачи, како што се класификација на слики, препознавање говор и обработка на природен јазик.

За време на фазата на обука, на моделот му се претставени многу примери на слики и нивните соодветни етикети и учи да ги препознава шаблоните и односите помеѓу сликите и етикетите. Откако моделот е обучен, може да земе нова слика како влез и да ја предвиди етикетата (т.е. животното) што одговара на таа слика.

Следниот пример покажува едноставен пример, како може да се предвидат некои трендови. Алгоритмите за регресија ќе бидат проучени подетално во идните поглавја, ова е само за демонстративни цели.

Го користиме алгоритамот за линеарна регресија кој во основа ни дава линеарна функција, која во математиката изгледа вака:

**f(x) = mx + b**

x - влез

f - излез

m - наклон

б - пресек со вертикалната оска

  Ако алгоритмот ни ги даде пресекот и наклонот, можеме да го процениме излезот за секој влез.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
# Генерирање на примероци
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 2, 5, 6])
```

```python
# Тренирање на моделот, употребувајќи ги познатите точки х и у
slope, intercept, _, _, _ = stats.linregress(x, y)
```

Симболите „долна црта“ ( _ ) означуваат дека на тие места ќе биде вметнати некои излези од функцијата, кои во моментов не се битни.
Функцијата *linregress* е само еден од можните начини да се пресмета линеарна регресија, алгоритам за кој повеќе ќе дознаете во Глава 8 од овој курс. Сепак, оние кои сакаат да дознаат повеќе детали за оваа функција, тоа можат да го направат [тука](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html#scipy-stats-linregress).

```python
# Генерирање на нови точки на оската х, и употреба на моделот за да се најдат нивните соодветни у координати
x_new = np.array([6, 7, 8])
y_pred = intercept + slope * x_new

```

```python
# Исцртување на оригиналните примероци (од кои моделот учеше) со сина боја
plt.scatter(x, y,color='blue')
# Исцртување на предвидените нови примероци со црвена боја
plt.scatter(x_new, y_pred, color='red')


# Прикажување на графикот
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Како што можеме да видиме, на алгоритмот му дадовме влез (х) и излез (у), од кои алгоритмот формираше модел („научил“ каква е зависноста на у од х ). Потоа на моделот му даваме нови влезови (кои не припаѓаат на оригиналното множество) и тој ни дава проекција за соодветните излези. 

## 6.3. Модел и инстанца
Модел е математичка претстава на систем или процес. Може да се користи за да се прават предвидувања или одлуки врз основа на податоци. Инстанца е специфична појава или пример за нешто, како што е одреден предмет или настан. Во контекст на машинското учење, инстанцата е специфично множество на влезни податоци што се користат за обука или тестирање на модел.

Пример за модел би бил фрлање петденарка. Во овој случај, моделот би бил математичка претстава за веројатноста при паѓањето на монетата да се добие петка или рис. Моделот би бил биномна дистрибуција со два можни исходи: петка (H - head) или рис (T - tail) со веројатност од 0,5 за секој исход. Инстанца на овој модел би бил конкретно набљудувано фрлање на монетата, во кое моделот би ја користел веројатноста од 0,5 за да го предвиди исходот од тоа фрлање.

Во нашиот претходен пример:

* **x** е влез,
* **y** е излез,
* **y_pred = intercept + slope * x_new** е моделот
* и примерот со влез **6** што резултира со излез **7** е инстанца

## 6.4. Регресија

Регресијата е метод за предвидување на нумеричка вредност врз основа на множество од влезни карактеристики. Пример е предвидување на цената на куќата врз основа на нејзината површина, бројот на спални соби и локацијата. Алгоритам за регресија зема множество на означени примери и произведува модел кој може да прави предвидувања за нови, неозначени примери. Во претходниот пример беше користен тип на алгоритам за регресија.

## 6.5. Класификација

Класификацијата е метод во машинското учење каде што моделот е обучен да додели одредена ознака (етикета) за класа на даден влез. Откривањето спам е пример за класификација, каде што моделот е обучен да предвиди дали е-поштата е спам или не. Друг пример би бил класификација на слики, каде што можеме на пример да препознаеме различни животни од слики.

Алгоритам за класификација зема множество означени примероци како влез и произведува модел кој потоа може да земе неозначен примерок и за него да додели ознака, или да додели број кој претставува веројатност за примерокот да се наоѓа во одредена класа.

И регресијата и класификацијата се надгледувани алгоритми, но регресијата се користи за предвидување на нумерички вредности, додека класификацијата се користи за предвидување категории или класи.

## 6.6. Ненадгледувано учење

Ненадгледуваното учење се занимава со проблеми во кои податочното множество се состои од неозначени примероци. Ова својство е она што го прави проблематично за многу практични примени. Отсуството на ознаки (етикети) кои го претставуваат посакуваното однесување на моделот воедно значи и отсуство на цврста референтна точка со која може да се оцени квалитетот и точноста на „наученото“. Ненадгледуваното учење може да се однесува понепредвидливо во споредба со другите природни методи на учење.

При ненадгледуваното учење моделот треба да работи сам за да открие шеми и информации кои претходно биле неоткриени. Но затоа, алгоритмите за ненадгледувано учење овозможуваат извршување на посложени задачи во споредба со надгледуваното учење. 

Реален пример за ненадгледувано учење е дете кое си игра со коцки и открива како тие се вклопуваат заедно за да изгради различни структури. На детето не му се даваат конкретни упатства или ознаки за коцките, туку тоа самото истражува и експериментира за да открие како може да се користат. Слично на тоа, дете кое си игра со различни играчки и открива како тие работат и што може да прави со нив е пример за ненадгледувано учење. Преку овие активности, детето може да научи и да ја разбере својата околина без директен надзор или водство.

Друг пример е кластерирањето, каде трговската компанија може да ги групира клиентите врз основа на нивните навики за купување. Алгоритмот ќе ја анализира историјата на купување на секој клиент и ќе ги групира во различни кластери врз основа на производите што тие имаат тенденција да ги купуваат и колку често ги купуваат. Компанијата потоа може да ги користи овие информации за да креира насочени маркетинг кампањи за секој кластер (група) клиенти.

Во основа, учењето без надзор е начинот на кој луѓето учат. Значи, ваквото однесување е почовечко, па е поблиску до вистинска вештачка интелигенција. Во реалните апликации, немаме секогаш означени податоци, а податоците не се секогаш категоризирани, исто така, не секој влезен податок има соодветен излез. Значи, во овие случаи користиме учење без надзор.
