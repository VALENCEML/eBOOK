# 8. Учење со надзор    

### Запознавање со учењето со надзор  

Учењето со надзор (учење со учител, анг. supervised learning) е вид на машинско учење каде што на компјутерот му се даваат примери од податоци со соодветни вредности (означени податоци), и тој ги користи тие информации за да научи како да го произведе точниот излез/вредност кога добива нови податоци. Во суштина, тоа е како на ученикот да му се покаже како да реши математички проблем, а потоа да вежба слични проблеми за да биде подобар во нивното решавање. При учење со надзор, компјутерот делува како ученик а означените податоци се како примерите што му се прикажуваат. Целта е компјутерот да прави точни предвидувања или одлуки врз основа на влезните податоци, исто како што ученикот треба да може да го примени она што го научил за решавање на нови проблеми. Концептот на надгледувано учење е прикажан на едноставен начин на сл. 1, каде што алгоритмот ја наоѓа зелената линија која ги одвојува податоците, со што ги класифицира според нивната вредност (X или O).

Пристапите кои користат учење со надзор дефинираат функција која ги пресликува влезните податоци во вредности/етикети (анг. labels). Оваа функција се нарекува **модел** или **хипотеза**, а задачата на алгоритмот за машинско учење е да ја најде најдобрата можна верзија на оваа функција со нагодување на нејзините **параметри**. Потоа, кога на компјутерот ќе му се дадат нови неозначени податоци (т.е. кога ги не му ги кажуваме нивните вредности), тој ги внесува овие податоци во хипотезата и произведува вредност наречена **предвидување**. Податоците што се користат за нагодување на моделот се нарекуваат **податоци за тренирање**, а новите неозначени податоци што се користат за тестирање на успехот на моделот се нарекуваат **податоци за тестирање** или **влезни податоци**. Откако ќе го добиеме предвидувањето на алгоритмот, го споредуваме со вистинската вредност на податоците за тестирање и ја оценуваме **грешката**. Врз основа на оваа грешка, одлучуваме дали моделот е задоволително точен или пак дали треба да се подобри. Целиот овој процес може да се види на сл. 2.

### Обележја и простор на обележја  

Начинот на кој алгоритмите за машинско учење ги градат овие модели е со испитување на **облежјата (анг. features)** на податоците за тренирање. Обележјата се карактеристики или атрибути на податоците што може да се користат за тие да се опишат. На пример, ако се обидувате да класифицирате различни животни, некои обележја можат да бидат нивната тежина, висина, боја, број на нозе итн. Ако се обидувате да опишете куќа, некои обележја можат да бидат големината (квадратурата), староста, бројот на соби, бројот на катови од куќата итн.  

Многу е важно да се напомене дека обележјата се предодредени вредности кои доаѓаат со податоците. Алгоритмите за машинско учење немаат влијание врз обележјата; напротив, тие треба да ги земат предвид токму обележјата на податоците за тренирање при градењето на моделот, а потоа и обележјата на податоците за тестирање при тестирањето на моделот.  

**Просторот на обележја (анг. feature space)** е просторот што е дефиниран со обележјата. Може да се замисли како график каде секоја оска претставува по едно обележје, па ако податоците имаат _n_ обележја, тогаш може да бидат претставени со n-димензионален координатен систем. Просторот на обележја се состои од сите можни точки во тој график што може да се креираат со комбинирање на вредности за секое обележје. Со други зборови, просторот на обележја се состои од сите можни комбинации на вредности на обележја кои опишуваат една податочна точка.  

### Класификација и регресија  

Двата најчести типа на проблеми кои можат да се појават во учењето со надзор се класификација и регресија, во зависност од природата на вредноста на податоците.  

* **Класификацијата** се користи за предвидување на **категоричка вредност** или **класа**, како што е на пример „куче“ или „мачка“. На компјутерот му се даваат примери на податоци за тренирање со соодветната ознака за класа и тој ги користи тие информации за да научи како да ги категоризира новите податоци во правилната класа. На пример, на компјутерот може да му се прикажат слики од различни животни и да се каже на која категорија припаѓа секое (на пр. „куче“, „мачка“, „птица“ итн.). Потоа, кога ќе се даде нова слика од животно, компјутерот мора да може да го класифицира во правилната категорија.  

* **Регресијата** се користи за предвидување на **континуална вредност**, како што е на пример цената на една куќа. На компјутерот му се даваат примери на податоци за тестирање со соодветни континуални вредности и тој ги користи тие информации за да научи како да предвидува нови континуални вредности. На пример, на компјутерот може да му се прикаже големината и локацијата на различни куќи и да му се кажат нивните цени. Потоа, кога ќе се даде големината и локацијата на нова куќа, компјутерот треба да може да ја предвиди нејзината цена.  

Накратко, класификацијата се користи за предвидување на ознаките на класите, а регресијата се користи за предвидување на континуални вредности.  

Во ова поглавје ќе испитаме неколку методи за класификација и регресија.    

## 8.1. Алгоритам на k најблиски соседи (k-NN)  

### 8.1.1. Дефиниција  

**Алгоритмот на k најблиски соседи (анг. k-Nearest Neighbors (k-NN))** е алгоритам за машинско учење што се користи за класификација. Тој е еден од наједноставните пристапи кој често се користи како репер во машинското учење.  

Замислете дека имате збир на означени податочни точки кои се слики од два различни типа/категории на животни (мачки и кучиња). k-NN работи така што ги наоѓа _k_-те податочни точки кои се наоѓаат најблиску до нова податочна точка и потоа ја доделува таа нова податочна точка на категоријата што е најчеста меѓу овие _„k“_ најблиски соседи. Овде _k_ е број кој исто така се нарекува **хиперпараметар**, кој треба да го изберете пред воопшто да го започнете алгоритмот.  

На пример, ако изберете _k=3_ и имате нова податочна точка (слика) што сакате да ја класифицирате како мачка или куче, k-NN ќе ги најде 3-те најблиски (т.е. најслични) податоци до таа нова точка. Ако 2 од тие 3 најблиски точки се означени како „мачка“ и 1 е означена како „куче“, k-NN ќе ја класифицира новата податочна точка како „мачка“. Ова значи дека k-NN ја користи идејата за „победа на мнозинството“ меѓу најблиските податочни точки за да ги класифицира новите податоци. Ова е причината зошто се смета за „мрзелив“ алгоритам за учење, бидејќи нема фаза на тренирање и едноставно ги користи означените податоци за да прави предвидувања. За разлика од алгоритмите што ќе бидат прикажани подоцна, k-NN не гради модел или хипотеза и веднаш ја класифицира секоја нова податочна точка.  

Алгоритмот ја одредува сличноста помеѓу две податочни точки со пресметување на **растојанието** меѓу нив во просторот на обележја. Растојанието помеѓу две точки _X_ и _Y_ во n-димензионален простор на обележја може да се пресмета како:  

$D(X,Y)=\sqrt{\sum_{i=1}^n |x_{(i)}-y_{(i)}|^{2}}$  

При спроведување на „гласањето“ кое го одредува мнозинството од соседите, постојат две опции:  
* **Униформниот пристап** им дава исто значење на сите _k_ соседи на новата податочна точка. Гласот на секој сосед има вредност 1.  
* **Пристапот базиран на растојание** им дава различно значење на _k_ соседите во зависност од нивната оддалеченост од новата податочна точка. При гласање, гласовите на секој сосед имаат вредност од $\frac{1}{D}$. Тоа значи дека поблиските соседи се поважни и повлијателни отколку оние што се подалеку.  

### 8.1.2. Пример со податочното множество за ириси  

Сега ќе ја демонстрираме употребата на k-NN за класификација на група цвеќиња од родот ирис. Ова е добро познато податочно множество од 1936 година, и се состои од 150 примероци на цветови од ирис од 3 различни видови: сетоса, вирџиника, и версиколор (по 50 од секој). Видовите на ирис се всушност класите на примероците. За секој цвет имаме и 4 други познати вредности: должина на чашка, ширина на чашка, должина на ливче, и ширина на ливче. Тоа се всушност обележјата, што значи дека имаме 4-димензионален проблем. Иако е стара и прилично мала, оваа база на податоци често се користи како едноставна алатка за демонстрација на алгоритми за класификација како што е k-NN.  

Во продолжение, ја увезуваме базата на податоци, ги земаме само првите 2 карактеристики (должина на чашка и ширина на чашка) за да можеме да ги визуелизираме податоците во 2-димензионален простор на обележја, и гледаме како k-NN алгоритмот ги класифицира цвеќињата во една од 3-те можни класи (видови).  

```python
# Import libraries and iris dataset:
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from sklearn.metrics import accuracy_score
iris = datasets.load_iris()

X = iris.data[:, :2] # only take the first two features
y = iris.target
print("Има вкупно",X.shape[0],"податоци. Секој од нив има по",X.shape[1],"обележја.")

# Create color maps for the three classes:
cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

n_neighbors = 15 # define number of neighbors k 
acc = [] # define array to store accuriacies

for weights in ['uniform', 'distance']: # try k-NN for both uniform and distance voting
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights) # create a Neighbours Classifier
    clf.fit(X, y) # fit the data
    pred = clf.predict(X) # predict values for all data
    acc.append(accuracy_score(pred,y)) # compare the actual class and the predicted class for all data
    print("Има вкупно",accuracy_score(pred,y,normalize=False),"точно класифицирани примероци со",weights,"гласање.")

#### The following part is complicated but only concerns the plotting. You don't have to analyze it. ####
    # Plot the results. Assign a color to each point in the coordinate system:
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # range on the x-axis
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # range on the y-axis
    h = .02  # step size in the mesh
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))    
   
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]) # predict the value of each datapoint in the plot
    
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xlabel("Должина на чашка")
    plt.ylabel("Ширина на чашка")
    plt.title("Класификација со 3 класи (k = %i, '%s' тежини)"
              % (n_neighbors, weights))
plt.show()
#### The complicated part about plotting ends here. ####

print("Точноста при униформно гласање е","{:.2f}".format(acc[0]),".\n")
print("Точноста при гласање базирано на растојание е","{:.2f}".format(acc[1]),".\n")
```

**Точноста (анг. accuracy)** е метрика што се користи во класификацијата за да се оцени колку добро алгоритмот ги класифицирал податоците. Таа едноставно го брои бројот на успешно класифицирани податоци и го дели со вкупниот број на податоци:  

$\text{ТОЧНОСТ}=\frac{\text{БРОЈ НА ТОЧНО КЛАСИФИЦИРАНИ ПОДАТОЦИ}}{\text{ВКУПЕН БРОЈ НА ПОДАТОЦИ}}$  

Таа секогаш има вредност помеѓу 0 (ништо не е погодено правилно) и 1 (сè е точно погодено).

Како што може да се види, точноста е нешто подобра кога се користи гласање базирано на растојание. Ова не е секогаш правило и резултатите се разликуваат од случај до случај, но генерално користењето на гласањето базирано на растојание придонесува за подобар и поробусен класификатор кој дава приоритет на непосредното соседство на податочната точка.  

Да се обидеме да го извршите горниот код со различни вредности за хиперпараметарот _k_ и да набљудувате како бројот на соседи влијае на точноста на алгоритмот:  

```python
acc_uni=[] # list for accuracies when uniform voting
acc_dis=[] # list for accuracies when distance voting
max_k=20 # maximum value for k to try (try different ones up to 150)

for k in range(max_k): 
    w = "uniform" # uniform voting
    clf = neighbors.KNeighborsClassifier(n_neighbors=k+1, weights=w) # create a Neighbours Classifier
    clf.fit(X, y) # fit the data
    pred = clf.predict(X) # predict values for all data
    correct = accuracy_score(pred,y,normalize=False) # calculate number of correctly classified samples
    score = accuracy_score(pred,y,normalize=True) # calculate accuracy
    print("Има вкупно",correct,"точно класифицирани примероци за",k+1,"-NN и при",w,"гласање (acc=","{:.2f}".format(score),").")
    acc_uni.append(score) # add accuracy to list for later plotting
    
    w = "distance" # distance voting
    clf = neighbors.KNeighborsClassifier(n_neighbors=k+1, weights=w) # create a Neighbours Classifier
    clf.fit(X, y) # fit the data
    pred = clf.predict(X) # predict values for all data
    correct = accuracy_score(pred,y,normalize=False) # calculate number of correctly classified samples
    score = accuracy_score(pred,y,normalize=True) # calculate accuracy
    print("Има вкупно",correct,"точно класифицирани примероци за",k+1,"-NN и при",w,"гласање (acc=","{:.2f}".format(score),").\n")
    acc_dis.append(score) # add accuracy to list for later plotting
    
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1,max_k+1),acc_uni)
plt.xlabel("k")
plt.ylabel("Точност")
plt.title("Униформно гласање")
plt.subplot(1, 2, 2)
plt.plot(range(1,max_k+1),acc_dis)
plt.xlabel("k")
plt.ylabel("Точност")
plt.title("Гласање базирано на растојание")
plt.show() 
```

## 8.2. Линеарна регресија  

### 8.2.1. Дефиниција  

**Линеарната регресија (анг. linear regression)** е статистичка метода која помага да се предвиди врската помеѓу две променливи вредности на една податочна точка: нејзиното обележје и нејзината вредност. Целта е да се најде права (позната како линија на регресија) која најдобро ги опишува или опфаќа дадените податоците, што добро се увидува ако податоците и линијата на регресија се нацртаат на ист график. Таа права потоа може да се користи за предвидување на идните вредности на слични податоци за кои е познато истото обележје. Се користи (очигледно) за решавање на проблеми од типот на регресија.  

Едноставно кажано, линеарната регресија ни помага да ја разбереме врската помеѓу две променливи и да правиме предвидувања врз основа на таа врска. Оваа врска е дадена со хипотезата на моделот:  

$y_{pred}=\theta_{0}+\theta_{1}x$  

Како што можете да видите, моделот не е ништо повеќе од равенка за права во координатен систем $(x,y)$. Овде, $x$ е обележјето на податочната точка и $y_{pred}$ е предвидената вредност на податочната точка. На пример, $x$ може да биде големина на куќа (во квадратни метри), а $y_{pred}$ може да биде предвидената цена на таа куќа. Линеарната регресија ни дава модел кој се обидува да ја најде врската помеѓу големината и вредноста на куќата. Компјутерот го прави ова со наоѓање на најдобрите можни вредности за **параметрите** $\theta_0$ и $\theta_1$.  

Ова нагодување на параметрите е всушност делот за машинско учење на алгоритмот. На компјутерот му е потребна база на податоци за обука со многу податочни точки во форма на [големина, цена]. Користејќи ги тие податоци, алгоритмот ја наоѓа правата што најдобро ги опишува пресметувајќи колку е оддалечена секоја точка од правата (колку поблиску - толку подобро). Правата што се приближува до податочните точки што е можно поблиску е најдобрата можна права (сл. 3).

Овие растојанија од податочните точки до правата ги претставуваат разликите помеѓу вистинската вредност на секој од податоците $y_{act}$ и вредноста $y_{pred}$ предвидена од алгоритмот (правата).  

### 8.2.2. Функција на цена и оптимални параметри  

Овие работи можеби нам ни се визуелно јасни и лесни за разбирање, но на компјутерот му треба да бидат преведени на математички јазик за да ги интерпретира и изврши. Затоа, велиме дека квалитетот на моделот е претставен со „средноквадратната грешка (анг. mean squared error - MSE) на сите податоци“ $J$:  

$J(\theta_0,\theta_1)=\frac{1}{2m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)^2=\frac{1}{2m}\sum_{i=1}^{m}(\theta_0+\theta_1x^i-y_{act}^i)^2$

Овде, $m$ е бројот на податочни точки во базата на податоци за тренирање. Колку е помала оваа вкупна грешка, толку е подобар моделот. Значи, алгоритмот всушност се обидува да ги пронајде параметрите $\theta_0$ и $\theta_1$ кои ја минимизираат функцијата $J$. Тие параметри ќе ја дадат равенката на најдобрата можна права за дадената база на податоци.  

Без да навлегуваме во премногу математички детали, алгоритмот го постигнува ова со тренирање на следниов начин:  

$\text{Иницијализација } \theta_0=\theta_1=0$  
$\text{За предодреден број на чекори или сѐ додека не се достигне конвергенција, повторувај го следново:}$  
$\{$  
&emsp;$\theta_0=\theta_0-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)$  
&emsp;$\theta_1=\theta_1-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)x^i$  
$\}$

Во овој случај, постигнување на конвергенција значи дека по некое време вредноста на $J$ престанува значително да се менува. Овде, $\alpha$ е хиперпараметар наречен **стапка на учење (анг. learning rate)**, со вредност помеѓу 0 и 1, и тој одредува колку брзо ќе се одвива тренирањето/учењето. Малите вредности на $\alpha$ ќе резултираат со бавно но сигурно учење, додека преголемата вредност може да го дестабилизира алгоритмот и да произведе безвредни и бескрајно големи вредности за параметрите $\theta_0$ и $\theta_1$.  

Кога ќе заврши горната процедура, ги добиваме **оптималните вредности** на параметрите и најдобрата можна права која ги опишува податоците. Следно, ако сме заинтересирани за цената на новата куќа, можеме да ја внесеме нејзината големина (нејзиното обележје) во нашиот модел, да видиме која ќе биде предвидената вредност, и да ја споредиме со вистинската вредност. За да оцениме, повторно ја користиме MSE.  

### 8.2.3. Пример со едно обележје

Следно, ќе видиме пример за користење на линеарна регресија за да се изгради модел кој го предвидува неделниот туристички профит (даден во илјади евра) на повеќе мали градови на Балканот како функција од нивното население. Населението на секој град е обележјето $x$, додека профитот на секој град е вредноста $y$.

По увезувањето, печатењето и визуелизирањето, податоците ќе ги поделиме на подмножества за тренирање и за тестирање и ќе го извршиме алгоритамот за линеарна регресија.  

```python
# Import libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = pd.read_csv('https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/profit.txt', header=None, names=['Population', 'Profit']) # open data file
print ("Преглед на податоците:\n",data.head())
print ("Статистика за податоците:\n",data.describe())

data.plot(kind='scatter', x='Population', y='Profit', figsize=(8,4));   # plot the data points
```

```python
X = data.iloc[:,0] # the features are contained in all rows, first column
y = data.iloc[:,1] # the values are contained in all rows, second column

# Split the data into traning and test subsets (ratio 9:1): 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/10, random_state=0) 
# Check the dataset and subsets dimensions: 
print('Големина на податочно множество: ', data.shape)
print('Големина на подмножество на обележја за тренирање: ', X_train.shape)
print('Големина на подмножество на обележја за тестирање: ', X_test.shape)
print('Големина на подмножество на вредности за тренирање: ', y_train.shape)
print('Големина на подмножество на вредности за тестирање: ', y_test.shape)

# Run the algorithm:
lr = LinearRegression()
lr.fit(X_train.values.reshape(-1,1),y_train) # train
# Get the coefficients: 
theta_0 = lr.intercept_ 
theta_1 = lr.coef_
print("Theta_0=","{:.2f}".format(theta_0))
print("Theta_1=","{:.2f}".format(theta_1[0]))
```

По тренирањето на алгоритмот со подмножеството за тренирање, гледаме дека равенката на правата е:  

$y_{pred}=-3.73+1.17x$  

Потоа ги пуштаме податоците за тестирање низ оваа равенка и ја пресметуваме средноквадратната грешка како индикатор за успехот на моделот (што помала - тоа подобро).  

```python
y_pred = lr.predict(X_test.values.reshape(-1,1)) # predict
print('Средноквадратна грешка:', metrics.mean_squared_error(y_test, y_pred)) # evaluate 
```

Средноквадратната грешка е 10,58, а нејзиниот квадратен корен е околу 3,25. Ако го споредиме ова со средната вредност (5.84) и стандардната девијација (5.51) на профитот на сите податоци (дадени погоре со методот data.describe()), ќе видиме дека оваа вредност на MSE е прифатлива и во согласност со податоците.  

Конечно, да ја нацртаме добиената права и да видиме како таа ги опфаќа податоците. Ќе додадеме и дополнителен и досега непознат град и ќе видиме што вели моделот за него.  

```python
x = np.linspace(data.Population.min(), data.Population.max(), 100)
f = theta_0 + theta_1 * x # this is the linear model for prediction with the parameters theta

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x, f, 'r') # plot the prediction line 
ax.scatter(data.Population, data.Profit) # plot the data points
ax.set_xlabel('Население (во илјади)')
ax.set_ylabel('Профит (во илјади)')
ax.set_title('Предвиден профит vs. Население')

prof = lr.predict([[20]]) # add a new town with a population of 20
print("На пример, според нашиот модел, град со популација од 20.000 жители би имал неделен профит од","{:.2f}".format(1000*prof[0]),"EUR.")
ax.scatter(20, prof, c='r'); # plot the town and it's profit value on the same figure 
```

Се разбира, ова не е совршено вклопување бидејќи податочните точки не се точно линеарно распределени и не можат никако сите да лежат на иста права. Но тие следат приближно линеарен тренд, така што велиме дека ова е речиси линеарен проблем и можеме да користиме линеарна регресија за него. Линеарната регресија може да се користи само за проблеми кои се линеарни или имаат трендови кои се блиску до линеарни.  

### 8.2.4. Пример со повеќе обележја  

Во реални проблеми, вредностите ретко зависат само од едно обележје. Профитот на еден град не зависи само од населението, туку и од многу други фактори (големина, локација, земја, клима, итн.). Слично на тоа, цената на куќата зависи од повеќе отколку само нејзината големина (бројот на соби, староста, градот, соседството итн.). Затоа линеарната регресија многу ретко се користи само со едно обележје, туку со повеќе.  

Сепак, без оглед на бројот на функции, алгоритмот воопшто не се менува. Множеството на податоци може да има $n$ обележја и на тој начин да биде n-димензионално. Ние тогаш не би можеле да го нацртаме или да го визуелизираме ако $n>3$, но сепак линеарната регресија ќе си функционира. Единствената разлика е во тоа што моделот повеќе нема да биде даден со 2-димензионална права, туку со n-димензионална хиперрамнина.  

Моделот сега е даден со:  

$y_{pred}=\theta_{0}+\theta_{1}x_1+...+\theta_{n}x_n$  

Ова е исто така и равенката на хиперрамнината. Тоа значи дека во принцип моделот има $n+1$ параметри.  

И самиот алгоритам е сличен:  

$\text{Иницијализација } \theta_0=\theta_1=...=\theta_n=0$  
$\text{За предодреден број на чекори или сѐ додека не се достигне конвергенција, повторувај го следново:}$  
$\{$  
&emsp;$\theta_0=\theta_0-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)$  
&emsp;$\theta_1=\theta_1-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)x_1^i$  
&emsp;...  
&emsp;$\theta_n=\theta_n-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)x_n^i$  
$\}$  

Средноквадратната грешка е исто така иста. Тоа значи дека линеарната регресија добро се скалира за проблеми со повеќе обележја и можеме да ја покажеме нејзината работа во следниот пример со две обележја за да можеме сè уште да ги визуелизираме податоците.  

Треба да изградиме модел кој ќе ја предвиди цената на една куќа во евра $y$ ако ја знаеме големината во квадратни стапки (обележје $x_1$) и бројот на спални соби (обележје $x_2$). Забележете како го користиме речиси истиот код како порано.

Сепак, постои една забележителна разлика: прво мора да ги **стандардизираме** податоците. Тоа значи дека треба да ја одземеме средната вредност на податоците од секоја податочна точка, а потоа да го поделиме резултатот со стандардното отстапување (девијација) на податоците. Како резултат на тоа, добиваме нови податоци кои ги содржат истите информации како и порано, но сега имаат средна вредност од 0 и стандардна девијација од 1. Ова мора да се направи кога се работи со повеќе обележја, бидејќи генерално тие се различно распределени во просторот. На пример, обележјето што ја означува големината на куќата може да се движи од 800 до 4500 квадратни стапки, но обележјето што го означува бројот на соби се движи само од 1 до 5. За да се избегне оваа нерамномерна распределба која може да предизвика грешки и бавно извршување на алгоритмот, податоците мора да се стандардизираат.  

```python
# Import libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = pd.read_csv('https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/prices.txt', header=None, names=['Size', 'Bedrooms', 'Price']) # open data file
print ("Преглед на податоците:\n",data.head())
print ("Статистика за податоците:\n",data.describe())

# Standardize the features in the multiple variable regression: 
data_s = (data - data.mean() ) / data.std()
print ("Стандардизирани податоци:")
print (data_s.head())
print (data_s.describe())

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data_s.Size, data_s.Bedrooms, data_s.Price)  # plot the data points
```

```python
X = data_s.iloc[:,:2] # the features are contained in all rows, first two columns
y = data_s.iloc[:,2] # the values are contained in all rows, third column

# Split the data into traning and test subsets (ratio 9:1): 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/10, random_state=0) 
# Check the dataset and subsets dimensions: 
print('Големина на податочно множество: ', data.shape)
print('Големина на подмножество на обележја за тренирање: ', X_train.shape)
print('Големина на подмножество на обележја за тестирање: ', X_test.shape)
print('Големина на подмножество на вредности за тренирање: ', y_train.shape)
print('Големина на подмножество на вредности за тестирање: ', y_test.shape)

# Run the algorithm:
lr = LinearRegression()
lr.fit(X_train,y_train) # train
# Get the coefficients: 
theta_0 = lr.intercept_ 
theta_1 = lr.coef_[0]
theta_2 = lr.coef_[1]
print("Theta_0=","{:.2f}".format(theta_0))
print("Theta_1=","{:.2f}".format(theta_1))
print("Theta_2=","{:.2f}".format(theta_2))
```

По тренирањето на алгоритмот со подмножеството за тренирање, гледаме дека равенката за 2-димензионалната рамнина е:  

$y_{pred}=-0.07+0.81x_1-0.03x_2$  

Потоа ги пуштаме податоците за тестирање низ оваа равенка и ја пресметуваме средноквадратната грешка како индикатор за успехот на моделот (што помала - тоа подобро).  

```python
y_pred = lr.predict(X_test) # predict
print('Средноквадратна грешка:', metrics.mean_squared_error(y_test, y_pred)) # evaluate 
```

Конечно, да ја нацртаме добиената рамнина и да видиме како таа ги опфаќа податоците. Ќе додадеме и дополнителна и досега непозната куќа и ќе видиме што вели моделот за неа.  

```python
x12 = np.linspace(data_s.Size.min(), data_s.Size.max(), 100)
x22 = np.linspace(data_s.Bedrooms.min(), data_s.Bedrooms.max(), 100)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data_s.Size, data_s.Bedrooms, data_s.Price, c='c')  # plot the data points
X12,X22 = np.meshgrid(x12,x22)

f = theta_0 + theta_1 * X12 + theta_2 * X22 # this is the linear model for prediction with the parameters theta

ax.plot_surface(X12, X22, f) # plot the prediction plane
ax.set_xlabel('Големина (кв. стапки) - стандардизирана')
ax.set_ylabel('Број на соби - стандардизирана')
ax.set_zlabel('Цена [EUR] - стандардизирана')
ax.set_title('Предвидена цена vs. Големина и број на соби')

new_house = data.iloc[-1,:] # take a new house (the last data point from the file for example: 1203 sq feet, 3 bedrooms)
new_house = (new_house - data.mean() ) / data.std() # standardize it
new_X = new_house[:2] # take only the features
price = lr.predict(new_X.values.reshape(1,-1)) # predict the value of this data point 
ax.scatter(new_X[0], new_X[1], price, c='r'); # plot the house and it's price on the same figure in red
```

Точката лежи точно на рамнината, но незначително се разликува од нејзината вистинска вредност во близина.  

Очигледно е дека рамнината не ги покрива сите точки совршено, што е невозможно бидејќи тие не се линеарно распределени. Но, таа успешно го доловува трендот на податоците кој е блиску до линеарен. Ова е најмногу што можеме да побараме од линеарната регресија кога ја користиме за решавање на реални и нелинеарни проблеми.

## 8.3. Логистичка регресија  

### 8.3.1. Дефиниција  

**Логистичка регресија (анг. logistic regression)** е статистички метод кој се користи во машинското учење за да се предвиди бинарен исход (т.е. два можни резултати, како да/не или точно/неточно) врз основа на една или повеќе влезни променливи. Како што дискутиравме претходно, ова е всушност процес на бинарна класификација.  

Замислете дека се обидувате да предвидите дали студентот ќе го положи или падне испитот врз основа на бројот на часови што ги поминал во учење. Можете да ги нацртате овие информации на график, со бројот на часови за учење на X-оската и исходот на положување на испитот на Y-оската (сл. 4).

Точките на графикот ги претставуваат поединечните студенти и нивните часови на учење и резултати од испитот. Забележете дека како што се зголемува бројот на часови за учење, така во главно се зголемува и веројатноста за полагање на испитот. Ова е местото каде што до израз доаѓа логистичката регресија - таа ни помага да ја најдеме најдобрата линија што ги раздвојува двата исхода што е можно подобро. Забележете дека линијата сега е закривена, за разлика од правата линија што ја имавме со линеарна регресија.  

Линијата може математички да се претстави со равенка, која ја зема влезната променлива (во овој случај, бројот на изучени часови) и дава веројатност за полагање на испитот. На пример, равенката за линијата на графикот погоре може да изгледа вака:  

$y_{pred}=\frac{1}{1+e^{-(\theta_0+\theta_1x)}}$  

Ова е моделот или хипотезата. Овде, $x$ се часовите за учење (обележје), а $\theta_0$ и $\theta_1$ повторно се параметрите што алгоритмот треба да ги пресмета за наоѓање на најдобрата можна линија. Конечно, $y_{pred}$ е вредноста што овој модел ќе ја предвиди - забележете дека се движи помеѓу 0 (ако експонентот на $e$ во заградите е многу негативен, именителот е позитивно бесконечен) и 1 (ако експонентот од $e$ во заградите е многу позитивен, именителот е 1). Затоа, моделот во логистичка регресија дава вредност на веројатност помеѓу 0 и 1 (0% и 100%) за тоа колку е веројатен исходот на проблемот. Но, бидејќи ова е бинарна класификација, велиме дека средината на овој опсег 0,5 (50%) е праг на одлука:  
* ако $y_{pred}>=0,5$, тогаш моделот предвидува 1 или ТОЧНО (TRUE);  
* ако $y_{pred}<0,5$, тогаш моделот предвидува 0 или ПОГРЕШНО (FALSE).  

Овој пристап се нарекува логистичка регресија иако се користи само за проблеми со класификација. Причината е тоа што моделот е повторно регресивен модел (изведува континуална вредност $y_{pred}$) како што може да се види на сл. 4. Но, токму воведувањето на прагот што го дефиниравме погоре го прави овој пристап класификациски пристап (излезите се 0 или 1). Името „логистичка“ доаѓа од типот на функцијата што се користи во моделот, која се нарекува логистичка функција.  

Како и кај линеарната регресија, логистичката регресија може да се користи и во проблеми со повеќе обележја, што е добро бидејќи повеќето реални проблеми имаат многу повеќе од едно обележје. Во случај на $n$ обележја (n-димензионален проблем), моделот е:  

$y_{pred}=\frac{1}{1+e^{-(\theta_0+\theta_1x_1+...+\theta_nx_n)}}$  

Без да навлегуваме во премногу математички детали, компјутерот ги оптимизира параметрите $\theta$ на сличен начин како што беше прикажано во линеарната регресија.  

Конечно, начинот на кој го оценуваме успехот на моделот на логистичка регресија е сличен како оној претходно претставен во k-NN: ги броиме сите правилно класифицирани примери од подмножеството на податоци за тестирање и тој број го делиме со вкупниот број на податоци во ова подмножество, со што се добива метриката за точност.    

$\text{ТОЧНОСТ}=\frac{\text{БРОЈ НА ТОЧНО КЛАСИФИЦИРАНИ ПОДАТОЦИ}}{\text{ВКУПЕН БРОЈ НА ПОДАТОЦИ}}$  

### 8.3.2. Пример за класификација со повеќе променливи

Да видиме пример каде имаме податоци за 100 студенти кои се пријавиле на Техничкиот универзитет. Ги знаеме нивните оценки на два матурски испити по математика и физика (ова се обележјата $x_1$ и $x_2$), и знаеме дали тие биле примени (1) или не (0) на Универзитетот (ова е класата). Да обучиме класификатор на логистичка регресија што гради модел и може да предвиди дали идните студенти ќе можат да се запишат (ќе бидат примени) врз основа на резултатите од нивните матурски испити.  

```python
# Import the libraries and the data:
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix

data = pd.read_csv('https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/exams.txt', header=None, names=['MathScore', 'PhysicsScore', 'Accepted?'])
print("All data:\n",data)
X = data.drop(columns=['Accepted?']) # x is all columns but 'Accepted?'
y = data['Accepted?'] # y is the 'Accepted?' column 
print("X:\n",X)
print("y:\n",y)

# Plot the data
positive = data[data['Accepted?'].isin([1])]  
negative = data[data['Accepted?'].isin([0])]
fig, ax = plt.subplots(figsize=(8,4))  
ax.scatter(positive['MathScore'], positive['PhysicsScore'], s=50, c='b', marker='o', label='Примен')  
ax.scatter(negative['MathScore'], negative['PhysicsScore'], s=50, c='r', marker='x', label='Не е примен')  
ax.legend()  
ax.set_xlabel('Резултат по математика')  
ax.set_ylabel('Резултат по физика');
```

Можеме да видиме дека податоците се прилично очекувани - оние ученици со повисоки резултати се генерално прифатени на Универзитетот. Но, за да видиме каде точно е разделбата, го тренираме моделот на логистичка регресија.  

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/10, random_state=0) # get train and test subsets
model = LogisticRegression() # create the classifier
model.fit(X_train, y_train) # train the classifier
# Return the coefficients theta: 
print("Theta_0 = ","{:.2f}".format(model.intercept_[0]))
print("Theta_1 = ","{:.2f}".format(model.coef_[0,0]))
print("Theta_2 = ","{:.2f}".format(model.coef_[0,1]))
```

По тренирањето на алгоритмот со подмножеството за тренирање, гледаме дека моделот е:  

$y_{pred}=\frac{1}{1+e^{-(-24.14+0.19x_1+0.20x_2)}}=\frac{1}{1+e^{(24.14-0.19x_1-0.20x_2)}}$  

Потоа ги пуштаме податоците за тестирање низ оваа равенка и ја пресметуваме точноста како индикатор за успехот на моделот.  

```python
y_pred = model.predict(X_test)
acc = metrics.accuracy_score(y_test,y_pred)
print("Точност=","{:.2f}".format(acc))
```

Следно, да ги претставиме овие резултати и на еден поинаков начин. Точноста дава само една вредност за квантифицирање на моделот, но ако користиме **матрица за конфузија**, можеме да добиеме повеќе информации. Матрицата за конфузија има 2 реда и 2 колони. Редовите ги претставуваат вистинските вредности на класите (0 во редот 0 и 1 во редот 1). Колоните ги претставуваат предвидените вредности на класите (0 во редот 0 и 1 во редот 1).  
* Бројот на локацијата (0,0) ги претставува **вистинските негативи (анг. true negatives - TN)**. Ова е бројот на податочни точки кои биле од класата 0 и биле правилно предвидени како класа 0.  
* Бројот на локацијата (0,1) ги претставува **лажните позитиви (анг. false positives - FP)**. Ова е бројот на податочни точки кои биле од класа 0 и биле погрешно предвидени како класа 1.  
* Бројот на локацијата (1,0) ги претставува **лажните негативи FN (анг. false negatives - FN)**. Ова е бројот на податочни точки кои биле од класа 1 и биле погрешно предвидени како класа 0.  
* Бројот во локацијата (1,1) ги претставува **вистинските позитиви TP (анг. true positives - TP)**. Ова е бројот на податочни точки кои биле од класа 1 и биле правилно предвидени како класа 1.  

```python
plot_confusion_matrix(model,X_test,y_test,cmap='Reds');
```

Како што можеме да видиме, во подмножеството за тестирање имало вкупно 10 податоци, 5 позитивни и 5 негативни. Сите 5 позитивни податоци биле правилно класифицирани (види десна колона), но има една грешка при класифицирањето на 5-те негативни податоци (види лева колона). Ова го потврдува претходно добиениот резултат на точност од 90%. Матрицата за конфузија е корисна алатка при визуелизација на точноста и многу поинформативна отколку само самиот резултат за точност, па затоа често се користи кај сите алгоритми за класификација.  

#### 8.3.2.1. Линија на одвојување  

Конечно, треба да се забележи уште една важна работа. Иако моделот на логистичка регресија повеќе не дава права, во него всушност се крие права (или во општ случај хиперрамнина). Експонентот на $e$ во логистичката функција е всушност равенка за права или рамнина што ги раздвојува податоците во просторот на обележја. За да го покажеме ова, ајде да ја нацртаме во просторот на обележја правата формирана од параметрите што ги добивме.  
$x_2 = -\frac{\theta_0}{\theta_2}-\frac{\theta_1}{\theta_2}x_1=\frac{24.14}{0.20}-\frac{0.19}{0.20}x_1$  

```python
x = np.linspace(data.MathScore.min(), data.MathScore.max(), 100)
f = -model.intercept_[0]/model.coef_[0,1] - model.coef_[0,0]*x/model.coef_[0,1]
fig, ax = plt.subplots(figsize=(8,4))  
ax.scatter(positive['MathScore'], positive['PhysicsScore'], s=50, c='b', marker='o', label='Примен')  
ax.scatter(negative['MathScore'], negative['PhysicsScore'], s=50, c='r', marker='x', label='Не е примен')  
ax.legend()  
ax.set_xlabel('Резултат по математика')  
ax.set_ylabel('Резултат по физика');
ax.plot(x, f, 'g', label='Линија на одвојување');
```

Оваа линија за одвојување (исто така наречена и **граница на одлука, анг. decision boundary**) секогаш ќе биде права, така што алгоритмот ќе прави грешки. Но, ако распределбата на податоци е речиси линеарна со некои исклучоци, тогаш логистичката регресија ќе работи добро (забележете ги сличностите со линеарната регресија во оваа смисла).  

### 8.3.3. Класификација при повеќе класи

Ова се прави со користење на таканаречениот метод „еден наспроти сите“ (сл. 5). Класификаторот што дава највисок резултат на веројатност „победува“, и податокот се класифицира во неговата класа.

## 8.4. Дрва на одлучување  

### 8.4.1. Дефиниција  

**Дрво на одлучување (анг. decision tree)** е тип на алгоритам за машинско учење што решава проблеми со разложување на комплексен проблем на помали потпроблеми полесни за решавање, а потоа носи одлуки врз основа на податоците достапни на секој чекор. Овој алгоритам не гради експлицитен модел за сите податоци однапред, и во оваа смисла се разликува од логистичката и линеарната регресија. Може да се користи и за класификација и за регресија, но полесно е да се разбере и да се прикаже во контекст на класификација, па тоа е она што ќе го направиме овде.  

Замислете дека сакате да го предвидите видот на овошјето што го имате врз основа на неговите боја и форма. Можете да започнете со поставување на серија прашања, како што се:  

Дали овошјето е меко?  
* а. Да:  
   * i. Дали овошјето е црвено?  
     * 1. Да: Плодот е јагода.  
     * 2. Не: Овошјето е банана.  
* б. Не:  
   * i. Дали овошјето е црвено?  
     * 1. Да: Плодот е калинка.  
     * 2. Не: Плодот е ананас.  
    
Секој чекор во процесот помага да се ограничат можните избори, сѐ додека не може да се направи конечно предвидување. Ова е слично на тоа како функционира дрвото на одлучување во машинското учење, каде што секој чекор е претставен со **јазол** во дрвото, а одлуките донесени во секој јазол водат во различни **гранки**. Конечното предвидување е направено во **лист**. Така, дрвото на одлучување може да се визуелизира како дијаграм сличен на дрво, каде што гранките ги претставуваат различните одлуки донесени на секој јазол, а листовите ги претставуваат конечните предвидувања.  

На сл. 6. е даден пример за дрво на одлучување кое се користи за класификација на овошје врз основа на неговите боја и облик.

Во овој пример, првата одлука е дали плодот е жолт или не. Ако е жолт, алгоритмот проверува дали е долг и слаб или не. Ако е така, овошјето се класифицира како банана ако лесно се лупи или како плантан ако не се лупи лесно. Ако плодот не е долг и слаб, тогаш се класифицира како јаболко. Ако пак овошјето не е жолто, алгоритмот проверува дали е портокалово или не. Ако е така, тогаш овошјето се класифицира како портокал ако е помало од грејпфрут или како грејпфрут ако не е. Ако не е во портокалова боја, тогаш плодот е класифициран како јаболко.  

Како што може да се види, навигацијата низ дрвото на одлучување е многу лесна бидејќи само треба да одговараме на прашања и да бираме гранки во зависност од нашите одговори, сѐ додека не стигнеме до лист. Токму вака изгледа фазата на тестирање, кога податоците од подмножеството за тестирање се внесуваат во дрвото и го „поминуваат“ во зависност од вредностите на нивните обележја сѐ додека не дојдат до лист што ја содржи нивната класа/вредност. Но, тешкото прашање е: како е изградено дрвото со помош на податоците за тренирање, т.е. како се избираат прашањата во јазлите?     

### 8.4.2. Нееднаквост и градење на дрвото

Дрвата за одлучување користат метрика наречена **Gini impurity (нечистотија, нееднаквост)** за да одлучат како да се поделат податоците, односно кои обележја и вредности да ги разгрануваат јазлите. Gini е мерка за тоа колку е „измешана“ една група на објекти. Замислете дека имате сад со M&M бонбони и сакате да ги поделите во две посебни чинии според бојата. Целта е да имате еден сад со сите сини бонбони, а друг сад со сите црвени бонбони. Колку повеќе се измешани бонбоните, толку е поголема оценката за нечистотија. Ако двете чинии се совршено поделени по боја, тогаш оценката за нечистотија ќе биде 0, што покажува дека бонбоните воопшто не се измешани. Во машинското учење, ја користиме нечистотијата за да процениме колку добро може да се подели множество од објекти во различни групи врз основа на одредени обележја. Целта е да се има низок резултат за нечистотија, што значи дека предметите се добро поделени во групи со слични обележја.  

Кога дрво се тренира, компјутерот го испробува секое обележје како корен. Ги разгранува сите варијанти и ја пресметува нечистотијата по тоа разгранување. Обележјето кое дава најниска оцена за нечистотијата е избрано како корен и е разгрането. Потоа, постапката се повторува повторно на сите следни јазли, на секоја длабочина и за сите обележја. Ова продолжува сè додека не достигнеме однапред одредена максимална длабочина на дрвото, или пак сѐ додека немаме повеќе јазли за разгранување (т.е. додека не дојдеме до сите листови). Добивањето на лист значи дека имаме нечистотија од 0, бидејќи сите податоци се јасно и правилно класифицирани во тој јазол.  

Овој пристап може да ви изгледа долг и досаден - испробувањето на сите обележја кај секој јазол звучи заморно! Но запомнете дека колку подлабоко одиме во дрвото, толку помалку податоци имаме на располагање бидејќи секој јазол ги анализира само податоците што ги задоволуваат гранките што довеле до него. Така, работите одат побрзо и компјутерот може да си дозволи да пробува различни обележја одново и одново.  

Следно, ќе пробаме неколку различни дрва на одлучување со базата на податоци за ирис за да го илустрираме сето она што го кажавме погоре. Како и досега, ќе ги увеземе податоците, ќе ги поделиме на подмножества за тренирање и тестирање, а потоа ќе ги обучиме дрвата за одлучување пред да ја тестираме нивната точност. Ќе испробаме 3 различни дрва за одлучување и ќе ги споредиме нивните точности: едно со максимална длабочина од 2, едно со максимална длабочина од 5, и едно без зададена максимална длабочина (т.е. ќе се разгранува сѐ додека не се достигнат сите листови).  

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the Iris dataset:
iris = load_iris()

# Create a DataFrame with the features and target:
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
print(df)

# Split the data into training and testing sets:
X_train,X_test,y_train,y_test = train_test_split(df[iris['feature_names']],df['target'],test_size=0.25,random_state=0)

# Fit a decision tree model with a maximum depth of 2:
clf1 = DecisionTreeClassifier(max_depth=2, random_state=0)
clf1.fit(X_train, y_train)
# Evaluate the accuracy of the model on the test data:
y_pred1 = clf1.predict(X_test)
print("Точност со максимална длабочина 2:", accuracy_score(y_test, y_pred1))

# Fit a decision tree model with a maximum depth of 5:
clf2 = DecisionTreeClassifier(max_depth=5, random_state=0)
clf2.fit(X_train, y_train)
# Evaluate the accuracy of the model on the test data
y_pred2 = clf2.predict(X_test)
print("Точност со максимална длабочина 5:", accuracy_score(y_test, y_pred2))

# Fit a decision tree model with a maximum depth of None (no limit on the depth):
clf3 = DecisionTreeClassifier(max_depth=None, random_state=0)
clf3.fit(X_train, y_train)
# Evaluate the accuracy of the model on the test data:
y_pred3 = clf3.predict(X_test)
print("Точност без зададена максимална длабочина:", accuracy_score(y_test, y_pred3)) 
```

Гледаме дека најдобрата точност е постигната со максимална длабочина од 5 (или можеби и порано од тоа), но максималната длабочина од 2 не секако е доволна. За да го илустрираме ова, ајде да го нацртаме дрвото со неговата максимална длабочина и да видиме колку длабоко оди.  

```python
import matplotlib.pyplot as plt
# Plot the tree with maximum depth (clf3):
plt.figure(figsize=(20, 20))
tree.plot_tree(clf3, filled=True, feature_names=df.columns)
plt.show()
# Left -> True; Right >- False
```

Дрвото прво го дели коренот според широчината на ливчето, поставувајќи го својот праг на одлука на 0,8 цм. Цветовите со помали широчини на ливчиња веднаш се класифицираат како ирис од класа 1 затоа што наидуваме на подмножество од податоците за тренирање што не содржи никакви други цвеќиња (т.е. има нечистотија на Gini од 0). Затоа, ова е лист. Остатокот од податоците потоа се дели според обележјето должина на ливчето, со праг од 4,95 цм, итн. Како што можеме да видиме, дрвото има длабочина од 4 пред да ги достигне сите лисни јазли со нечистотија од 0, така што тоа е максималната длабочина.  

Ова е една од големите предности на дрвата на одлучување - можеме јасно да видиме и да разбереме зошто тие ги класифицираат податоците на начинот на кој тоа го прават. Сè што треба да направиме е да го нацртаме дрвото и да ги следиме јазлите и гранките за секој податок. Затоа стеблата за одлучување се нарекуваат „модел на бела кутија“ (анг. white-box model) во смисла на „проѕирна кутија“.  

Можеме да видиме и дека нечистотијата се намалува како што одиме сѐ подлабоко и работите стануваат сѐ појасни - започнуваме со нечистотија од 0,665 во коренот и завршуваме со нечистотија од 0 кај сите листови. Забележете и како започнуваме со вкупно 112 примероци (податоците за тренирање се 3/4 од оригиналната база на податоци од 150 цвеќиња), но бројот на примероци се намалува како што одиме подлабоко во дрвото.  

Следно, да ја видиме матрицата за конфузија за подмножеството за тестирање на ова дрво, а исто така и на дрвото со максимална длабочина од 2, кое имаше помал резултат за точност.   

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf3,X_test,y_test,cmap='Reds');
plot_confusion_matrix(clf1,X_test,y_test,cmap='Reds');
```

Од 38 примероци што беа користени за тестирање на дрвото со максимална длабочина, само еден бил погрешно класифициран (предвидена класа 2, но всушност бил класа 1). Дрвото пак со максимална длабочина од 2 направи уште 3 други грешки.

Забележете дека зголемувањето на максималната длабочина на дрвото генерално води до подобри перформанси кај податоците за тренирање, но може да резултира со **претренирање (анг. overfitting, учење на памет)** (т.е. моделот е премногу сложен и не генерализира добро на новите податоци). Добра практика е максималната длабочина да се нагоди така за да се најде најдобрата рамнотежа помеѓу точноста и генерализацијата.  

### 8.4.3. Поголем пример    

Следно, да пробаме пример со поголема база на податоци, со повеќе обележја и примероци. Тој содржи здравствени и физиолошки информации (обележја) за одреден број на пациенти и информации за тоа дали им е дијагностициран дијабетес или не (класа 1 или 0). Ако обучиме добар модел за класификација, можеме да го искористиме за да предвидиме или барем да помогнеме во дијагностицирање на дијабетес кај нови пациенти во иднина.  

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the dataset:
data = pd.read_csv("https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/diabetes.csv")

# Split the data into features and target:
X = data.drop("Outcome", axis=1)
y = data["Outcome"]

# Split the data into training and testing sets:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train a decision tree classifier:
dtree = DecisionTreeClassifier(max_depth=None, random_state=0)
dtree.fit(X_train, y_train)

# Predict the outcome of test data:
y_pred = dtree.predict(X_test)

# Calculate the accuracy of the model:
accuracy = accuracy_score(y_test, y_pred)
print("Точност: ","{:.2f}".format(accuracy))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(20, 20))
tree.plot_tree(dtree, filled=True, feature_names=X.columns)
plt.show()
# Left -> True; Right >- False
```

Дрвото сега е многу поголемо и достигнува максимална длабочина од 16. Да видиме како точноста на тестирањето се менува при различни нивоа на длабочина.  

```python
acc_train = [] # list for all the training accuracies
acc_test = [] # list for all the testing accuracies

for i in range(16): # try depths from 1 to 16
    dtree = DecisionTreeClassifier(max_depth=i+1, random_state=0) # generate tree with appropriate depth
    dtree.fit(X_train, y_train) # train tree
    y_pred_train = dtree.predict(X_train) # test tree on training data
    y_pred_test = dtree.predict(X_test) # test tree on testing data
    accuracy_train = accuracy_score(y_train, y_pred_train) # calculate training accuracy
    accuracy_test = accuracy_score(y_test, y_pred_test) # calculate testing accuracy
    acc_train.append(accuracy_train) 
    acc_test.append(accuracy_test) 

# Plot accuracies for different depths:
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1,17),acc_train)
plt.xlabel("Максимална длабочина")
plt.ylabel("Точност")
plt.title("Зависност на точноста при тренирање од максималната длабочина")
plt.subplot(1, 2, 2)
plt.plot(range(1,17),acc_test)
plt.xlabel("Максимална длабочина")
plt.ylabel("Точност")
plt.title("Зависност на точноста при тестирање од максималната длабочина")
plt.show() 
```

Очекувано, ако го тестираме дрвото на податоците за тестирање (кои веќе ги видело при тренинзите), прецизноста се подобрува со длабочината, па дури и достигнува 100%.  

Но, интересно е да се види дека при тестирање на невидени податоци, најдобрата точност не се јавува за најдлабокото дрво. Наместо тоа, ова се случува на помала длабочина од 10. Причината за ова е што подлабоките дрва имаат тенденција да се претренираат, односно тие „учат на памет“ кога бараат релација помеѓу обележјата и вредноста/класата. Реалниот еквивалент на ова би бил студент кој учи прашања и одговори напамет; ако му се постават истите тие прашања, совршено ќе ги знае одговорите. Но ако му се постави кое било друго прашање поврзано со материјалот но различно од она што го научил, ќе се мачи затоа што ги научил прашањата и одговорите без да ги разбере.  

Како што рековме претходно, сакаме добра рамнотежа помеѓу точноста и генерализацијата, па затоа треба да ја провериме точноста за различни вредности на хиперпараметрите (што во овој случај е максималната длабочина на дрвото) пред да се решиме за најдобрата.
