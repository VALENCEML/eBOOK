Për më tepër, nëse doni të dini statistikat e përdorimit të platformës së mediave sociale, kompjuteri do t'i konsiderojë **FB** dhe **Facebook** si dy platforma të ndryshme, megjithëse ju e dini që këto të dyja duhet të bashkohen. (E njëjta gjë për Insta/Instagram, ose YT/YouTube...). Kjo na nxit që të insistojmë në të dhëna të qëndrueshme, ndoshta mund të përgatisim listat e opsioneve të mundshme edhe para se të fillojmë digjitalizimin e të dhënave. Le të themi se kemi rënë dakord për rregullat e mëposhtme:
- gjinia do të përfaqësohet me shkronjën e madhe M ose F
- koha do të përfaqësohet nga një numër i plotë në minuta (pa njësi, vetëm një numër)
- platforma do të përfaqësohet nga një varg shkronjash, që gjenden ekskluzivisht në këtë listë: Facebook/Instagram/YouTube/TikTok/WhatsApp/WeChat.

Pas gjithë këtij shqyrtimi, më në fund mund të krijoni grupin e të dhënave:

```python
friends = pd.DataFrame(columns=['Name','Gender', 'Time Daily [min]', 'Platform'])
```

Grupi i të dhënave quhet ***friends***, dhe tani është bosh. Për të shtuar rreshtin e parë në grupin e të dhënave, mund të përdoret komanda 'append':

```python
friends = friends.append({'Name': 'Zdravko','Gender': 'M', 'Time Daily [min]': '60',  'Platform': 'Facebook'}, ignore_index=True)
friends
```

Komanda 'append' mund të mbështesë shtimin e shumë rreshtave në grupin e të dhënave në të njëjtën kohë, megjithatë, dikush duhet t'i shtypë ato me dorë:

```python
friends = friends.append(pd.DataFrame({
                'Name': ['Aleksandra','Ivan'],
                'Gender': ['F','M'], 
                'Time Daily [min]': ['90','120'],  
                'Platform': ['Instagram','TikTok']
                }) , ignore_index=True)
friends
```

Një mënyrë më e shpejtë dhe më e përshtatshme për të shtuar vetëm një rresht është sintaksa e mëposhtme:

```python
friends.loc[len(friends.index)] = ['Artim','M', '30', 'Facebook']
friends
```

Mund të shkruani edhe një kod të vogël me cikël i cili mund t'ju ndihmojë në futjen gjysmë automatike të të dhënave. Provoni dhe përfundoni listën e mësipërme ose shtoni disa nga miqtë tuaj në grupin e të dhënave. Shtoni të paktën 4 persona, për të marrë një grup të dhënash me të paktën 8 të dhëna (të rëndësishme për demonstrimet e ardhshme, në seksionin 3.3.)

```python
keepgoing = True  #initialisation
while keepgoing:  # perform all following commands while keepgoing is True
  Name = input("Name?")
  Gender = input("Gender? [M/F]")
  TimeD = input("Average time spent daily? [in minutes]")
  Platform = input("Social media platform? [Facebook/Instagram/YouTube/TikTok/WhatsApp/WeChat]")
  AnotherRecord = input("Make another entry? [y/n]")
  friends.loc[len(friends.index)] = [Name, Gender, TimeD, Platform]
  if AnotherRecord == 'n':
    keepgoing = False # when the user enters 'n', that breaks the loop
friends
```

Tani ju keni grupin tuaj të të dhënave. Do të mësoni rreth analizimit dhe vizualizimit të të dhënave në kapitujt e ardhshëm, por tani për tani, vetëm një sugjerim i vogël:


```python
from matplotlib import pyplot as plt
counts = friends.Platform.value_counts()
plt.pie(counts, labels=friends.Platform.value_counts().index.tolist())
plt.title("Social media platforms of my friends") 
```

Seti juaj i të dhënave ekziston në një kornizë të dhënash (një lloj ndryshoreje shumëdimensionale) në memorien e punës të kompjuterit. Kur të largoheni nga ky sesion, të gjitha të dhënat do të humbasin. Nëse dëshironi të ruani dhe shkarkoni të dhënat për përdorim në të ardhmen, kjo mund të bëhet me komandat e mëposhtme:

```python
from google.colab import files
friends.to_csv('friends.csv', encoding = 'utf-8-sig', index=False)
files.download('friends.csv')
```

Pas kësaj, grupi i të dhënave ruhet lokalisht në diskun e kompjuterit dhe në sesionin tuaj të ardhshëm mund ta importoni atë dhe të vazhdoni të punoni me të.

*Praktika e mëtejshme*: përpiquni të krijoni grupin tuaj të të dhënave, me parametra të tjerë (disa ide: muajt e lindjes, ngjyrat e syve ose flokëve, zhanri i preferuar i muzikës...). Kryeni një anketë midis miqve tuaj, krijoni grupin e të dhënave prej tij dhe ruajeni për përdorim në të ardhmen! Gjeni skedarin në kompjuterin tuaj, hapeni (çdo redaktues teksti mund të hapë skedarë csv, ose mund të përdorni një redaktues fletëllogaritëse, si Excel ose Calc) dhe shikoni strukturën e të dhënave brenda. ### 3.3.3. Fshirja (disa prej) të dhënave

Ndonjëherë, pjesë të kornizës së të dhënave nuk nevojiten më, ose duhet të përjashtohen nga analiza e mëtejshme. Një mënyrë për ta bërë këtë është të zgjidhni pjesën përkatëse të kornizës së të dhënave dhe ta caktoni atë në një kornizë tjetër (të re), por në disa raste zgjidhja më e thjeshtë është të fshini të dhënat e panevojshme.



```python
friends2 = friends.copy()   # don't destroy the original data set, make a new one
friends2
```

Për fshirjen e pjesshme të të dhënave, përdoret operatori *drop*.
Fshirja e një të dhëne specifike (indeksi i njohur):

```python
friends2.drop([5])
```

Vëzhgoni se si ruhen indekset origjinale. Kompleti i të dhënave nuk ri-indeksohet pas heqjes së një të dhëne.
Fshirja e shumë të dhënave:

```python
friends2.drop([0, 3, 7])
```

Vini re se rreshti me indeks=5 që kemi fshirë më sipër është kthyer në të dhëna.  Kjo për shkak se rezultati i mësipërm u krye në një kopje të sigurt të kornizës së të dhënave, dhe jo në vetë kornizën e të dhënave. Meqenëse i keni kopjuar tashmë të dhënat origjinale, mund të detyroni fshirjen e përhershme në vetë kornizën e të dhënave, duke përdorur operatorin *inplace*.

```python
friends2.drop([0, 3, 7],inplace=True)
friends2 # note that now the result must be explicitly printed, to see it
```

Fshirja sipas ndonjë kushti (indeksi i panjohur):

```python
friends2.drop(friends2[friends2['Gender']=='F'].index, inplace=True)   # remove all females from the data 
friends2
```

Heqja e kolonave:

```python
friends2 = friends.copy()   # bring back the data
friends2
```

Për të fshirë një kolonë duhet të specifikohet parametri *axis=1*. Mund të specifikohet edhe gjatë fshirjes së rreshtave (*axis=0*), por nuk është i detyrueshëm, sepse nëse nuk specifikohet, supozohet se fshihen rreshtat.

### 3.3.4. Kombinimi i të dhënave

Në disa situata, të dhënat nga burime të shumta duhet të kombinohen në një grup të dhënash. Për të simuluar atë skenar, ekzekutoni kodin në vijim:


```python
friends1 = friends.copy()  
friends2 = friends.copy()  
friends1.drop([0, 2, 5],inplace=True)
friends2.drop([1, 3, 7],inplace=True)
friends1
```

```python
friends2
```

Tani ju keni dy grupe të ndryshme të dhënash. Operatori *concat* i bashkon këto dy grupe në një.

```python
friends_all = pd.concat([friends1, friends2])
friends_all
```

Vini re tani se si është shkatërruar rendi i indekseve dhe grupi i ri i të dhënave përmban indekse jo unike. Ky në përgjithësi nuk është një problem për operacionet e të dhënave, por duhet të jeni shumë të kujdesshëm, sepse tani të dhënat e shumta mund të kenë të njëjtin indeks.
Nëse dëshironi që grupi i ri i të dhënave të rivendosë indekse të reja, të renditura, atëherë duhet të përdoret çelësi *ignore_index*.

```python
friends_all = pd.concat([friends1, friends2], ignore_index = True)
friends_all
```

Ose, nëse doni të ruani indekset e vjetra (ndoshta keni zbatuar disa analiza që funksionojnë me ta), por megjithatë duhet që çdo e dhënë të jetë e qasshme në mënyrë unike, atëherë mund të shtoni një indeks tjetër, me një vlerë hierarkike më të lartë se ajo e vjetra.

```python
friends_all = pd.concat([friends1, friends2], keys=['fr1','fr2'])
friends_all
```

Tani, çdo e dhënë është e qasshme në mënyrë unike dhe mund t'i caktohet një vlerë unike:

```python
friends_all.loc['fr1'].loc[4,'Platform'] = 'TikTok'
friends_all.loc['fr2'].loc[4,'Platform'] = 'Instagram'
friends_all
```

Po grupet e të dhënave që kanë struktura të ndryshme? Le të supozojmë se keni dy grupe të dhënash, njëra ka të bëjë me praninë e rrjeteve sociale dhe tjetra ka të bëjë me rezultatet e testimit. Dy grupe të dhënash kanë disa nga të dhënat që mbivendosen, por të tjerat janë të ndryshme. Për të ilustruar një skenar të tillë, së pari, ekzekutoni ndryshimet e mëposhtme në *friends2*:

```python
friends2['Time Daily [min]'] = friends2['Time Daily [min]'].apply(pd.to_numeric)
friends2 = friends2.rename(columns={'Time Daily [min]':'Test Score'}) # renaming the column
friends2['Test Score'] = friends2['Test Score'] + 48   # changing the values in column, to be different from the previous (just for visual appearance)
friends2
```

Tani, kombinimi i dy grupeve të të dhënave rezulton në:

```python
friends_all = pd.concat([friends1, friends2])
friends_all
```

Vini re *NaN*'s. Parametrat që nuk janë të disponueshëm, në grupin e të dhënave të bashkuara përfaqësohen nga *NaN*.
Ky rezultat është disi i përdorshëm, por për të bashkuar vërtetë të dhënat, kërkohen disa hapa të tjerë. Së pari, le të gjejmë të dhënat që ekzistojnë në të dy grupet fillestare të të dhënave.

```python
friends_common = friends1.merge(friends2,on=['Name','Gender','Platform'])  # assume that Name, Gender and Platform would be same in both data sets, because those are the same people.
friends_common
```

Pastaj, le t'i fshijmë ato të dhëna nga grupi i të dhënave të bashkuara. Një nga një, të gjithë emrat nga *friends_common* kontrollohen në *friends_all* dhe të gjitha të dhënat me atë emër fshihen.

```python
for i in range (len(friends_common.index)):   
  friends_all.drop(friends_all[friends_all['Name']==friends_common.loc[i,'Name']].index, inplace=True)  # try to analyze and understand this code, multiple concepts of selecting, deleting and assigning are used in the same row!
friends_all
```

Në fund, emrat e përbashkët duhet t'i shtohen pjesës tjetër të të dhënave:

```python
friends_all = pd.concat([friends_all, friends_common],ignore_index=True)
friends_all
```

Të dhënat janë bashkuar plotësisht dhe për njerëzit që ekzistojnë në të dy grupet e të dhënave.

*Ushtrimi:* nga grupi i të dhënave *friends_all* zgjidhni të gjitha të dhënat që nuk përmbajnë të dhëna për 'Time Daily [min]', në një kornizë të veçantë të dhënash. Më pas vendosni rezultatet e testit si kohë (duke riemërtuar dhe fshirë kolonat) dhe shtoni ato të dhëna përsëri në grupin e të dhënave *friends_all*, në fund të tabelës.
