### Karakteristike i prostor karakteristika

Način na koji algoritmi mašinskog učenja grade ove modele je pregledavanje **karakteristika** podataka za obuku. Karakteristike su osobine ili atributi podataka koji se mogu koristiti za opisivanje podataka. Na primer, ako pokušavate klasifikovati različite životinje, neke karakteristike mogu biti njihova težina, visina, boja, broj nogu itd. Ako pokušavate opisati kuću, neke karakteristike mogu biti veličina, starost, broj soba, broj spratova kuće itd.

Veoma je važno napomenuti da su karakteristike unapred određene vrednosti koje dolaze sa podacima. Algoritmi mašinskog učenja nemaju uticaja na karakteristike; naprotiv, moraju uzeti u obzir karakteristike podataka za obuku prilikom izrade modela, a zatim karakteristike test podataka prilikom testiranja modela.

**Prostor karakteristika** je prostor koji definišu karakteristike. Može se zamisliti kao grafikon u kojem svaka osa predstavlja jednu karakteristiku, pa ako podaci imaju n karakteristika, tada se mogu predstaviti n-dimenzionalnim koordinatnim sistemom. Prostor karakteristika sastoji se od svih mogućih tačaka u tom grafikonu koje se mogu kreirati kombinovanjem vrednosti za svaku karakteristiku. Drugim rečima, prostor karakteristika je sve moguće kombinacije vrednosti karakteristika koje opisuju podatke.

### Klasifikacija i regresija

Dve najčešće vrste problema koji se mogu javiti u nadgledanom učenju su klasifikacija i regresija, u zavisnosti od prirode vrednosti podataka.

- **Klasifikacija** se koristi za predviđanje kategorskog obeležja ili klase, poput "pas" ili "mačka". Računaru se pokazuju primeri podataka za obuku sa odgovarajućom klasnom oznakom, i on koristi te informacije kako bi naučio kako kategorisati nove podatke u ispravnu klasu. Na primer, računaru se mogu pokazati slike različitih životinja i reći kojoj kategoriji svaka pripada (npr. "pas", "mačka", "ptica" itd.). Zatim, kada dobije novu sliku životinje, računar mora biti sposoban da je klasifikuje u ispravnu kategoriju.

- **Regresija** se koristi za predviđanje **kontinualne vrednosti**, poput cene kuće. Računaru se pokazuju primeri podataka za obuku sa odgovarajućom kontinuiranom vrednošću, i on koristi te informacije kako bi naučio kako predviđati nove kontinualne vrednosti. Na primer, računaru se može pokazati veličina i lokacija različitih kuća i reći im cene. Zatim, kada dobije novu veličinu i lokaciju kuće, računar bi trebao moći da predvidi njenu cenu.

Ukratko, klasifikacija se koristi za predviđanje klasnih oznaka, a regresija se koristi za predviđanje kontinuiranih vrednosti.

U ovom poglavlju proučićemo nekoliko metoda za klasifikaciju i regresiju.

## 8.1. Algoritam k najbližih suseda (k-NN)

### 8.1.1. Definicija

**Algoritam k najbližih suseda (k-nearest neighbors)** je algoritam mašinskog učenja koji se koristi za klasifikaciju. To je jedan od najjednostavnijih pristupa, često se koristi kao referentna tačka.

Zamislite da imate skup označenih podataka koji su slike dve različite vrste/kategorije životinja (mačaka i pasa). K-NN radi tako što pronalazi k najbližih podataka novoj tački podataka, a zatim dodeljuje novu tačku podataka kategoriji koja je najčešća među "k" najbližim tačkama. Ovde je k broj koji se takođe naziva **hiperparametar**, koji morate odabrati pre pokretanja algoritma.

Na primer, ako odaberete k=3 i imate novu tačku podataka (sliku) koju želite da klasifikujete kao mačku ili psa, k-NN će pronaći 3 najbliže (tj. najličnije) podatke novoj tački u označenim podacima. Ako su 2 od tih 3 najbližih tačaka označene kao "mačka" i 1 kao "pas", k-NN će klasifikovati novu tačku podataka kao "mačka". To znači da k-NN koristi ideju "većina pobeđuje" među najbližim podacima za klasifikaciju novih podataka. Zato se smatra "lenjim" algoritmom učenja, jer nema fazu obuke i jednostavno koristi označene podatke za pravljenje predikcija. Za razliku od algoritama koji će biti prikazani kasnije, k-NN ne gradi model ili hipotezu i odmah klasifikuje svaku novu tačku podataka.

Algoritam određuje sličnost između dve tačke podataka izračunavanjem **udaljenosti** između njih u prostoru karakteristika. Udaljenost između dve tačke X i Y u n-dimenzionalnom prostoru karakteristika može se izračunati kao:
$D(X,Y)=\sqrt{\sum_{i=1}^n |x_{(i)}-y_{(i)}|^{2}}$

Prilikom sprovođenja "glasanja" koje određuje većinu suseda, postoje dve opcije:

- **Uniformni** pristup daje istu važnost svim k susedima nove tačke podataka. Svaki susedov glas ima vrednost 1.
- **Pristup na osnovu udaljenosti** daje različitu važnost k susedima u zavisnosti od njihove udaljenosti od nove tačke podataka. Prilikom glasanja, svaki susedov glasima vrednost $\frac{1}{D}$. To znači da su bliži susedi važniji i uticajniji od onih koji su dalje.



### 8.1.2. Primer sa skupom podataka Iris

Sada ćemo demonstrirati upotrebu k-NN algoritma za klasifikaciju nekih cvetova iz roda Iris. Ovo je poznati skup podataka iz 1936. godine, koji se sastoji od 150 uzoraka iris cvetova iz 3 različite vrste: setosa, virginica, versicolor (50 od svake). To su klase uzoraka. Za svaki cvet postoje još 4 poznate vrednosti: dužina čašičnog listića, širina čašičnog listića, dužina latice i širina latice. To su karakteristike, što znači da imamo problem sa 4 dimenzije. Iako je star i prilično mali, ovaj skup podataka se često koristi kao jednostavan alat za demonstraciju algoritama za klasifikaciju, kao što je k-NN.

U nastavku uvozimo skup podataka, uzimamo samo prve 2 karakteristike (dužinu čašičnog listića i širinu čašičnog listića) kako bismo mogli vizualizovati podatke u dvodimenzionalnom prostoru karakteristika, i isprobavamo k-NN algoritam za klasifikaciju cvetova u jednu od 3 moguće klase (vrste). 

```python
# Import libraries and iris dataset:
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from sklearn.metrics import accuracy_score
iris = datasets.load_iris()

X = iris.data[:, :2] # only take the first two features
y = iris.target
print("There are",X.shape[0],"data points. Each has",X.shape[1],"features.")

# Create color maps for the three classes:
cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

n_neighbors = 15 # define number of neighbors k 
acc = [] # define array to store accuriacies

for weights in ['uniform', 'distance']: # try k-NN for both uniform and distance voting
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights) # create a Neighbours Classifier
    clf.fit(X, y) # fit the data
    pred = clf.predict(X) # predict values for all data
    acc.append(accuracy_score(pred,y)) # compare the actual class and the predicted class for all data
    print("There are",accuracy_score(pred,y,normalize=False),"correctly classified samples with",weights,"voting.")

#### The following part is complicated but only concerns the plotting. You don't have to analyze it. ####
    # Plot the results. Assign a color to each point in the coordinate system:
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # range on the x-axis
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # range on the y-axis
    h = .02  # step size in the mesh
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))    
   
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]) # predict the value of each datapoint in the plot
    
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xlabel("Sepal length")
    plt.ylabel("Sepal width")
    plt.title("3-Class classification (k = %i, weights = '%s')"
              % (n_neighbors, weights))
plt.show()
#### The complicated part about plotting ends here. ####

print("The accuracy when using uniform voting is","{:.2f}".format(acc[0]),".\n")
print("The accuracy when using distance voting is","{:.2f}".format(acc[1]),".\n")
```

Tačnost (accuracy) je metrika koja se koristi u klasifikaciji kako bi se procenilo koliko dobro je algoritam klasifikovao podatke. Jednostavno broji broj uspešno klasifikovanih tačaka podataka i deli ga sa ukupnim brojem tačaka podataka:

$\text{Tačnost}=\frac{\text{BROJ TAČNO KLASIFIKOVANIH TAČAKA PODATAKA}}{\text{BROJ SVIH TAČAKA PODATAKA}}$

Ima vrednost između 0 (ništa nije tačno pogodjeno) i 1 (sve je tačno pogodjeno).

Kako se može videti, tačnost je nešto bolja kada se koristi glasanje na osnovu udaljenosti. Ovo nije uvek pravilo i rezultati se razlikuju od slučaja do slučaja, ali uopšteno, korišćenje glasanja na osnovu udaljenosti doprinosi boljem i robusnijem klasifikatoru koji prioritizuje neposredno susedstvo tačke podataka.

Takođe, trebali biste probati pokrenuti gornji kod sa različitim vrednostima za hiperparametar k i posmatrati kako broj suseda utiče na tačnost algoritma:

```python
acc_uni=[] # list for accuracies when uniform voting
acc_dis=[] # list for accuracies when distance voting
max_k=20 # maximum value for k to try (try different ones up to 150)

for k in range(max_k): 
    w = "uniform" # uniform voting
    clf = neighbors.KNeighborsClassifier(n_neighbors=k+1, weights=w) # create a Neighbours Classifier
    clf.fit(X, y) # fit the data
    pred = clf.predict(X) # predict values for all data
    correct = accuracy_score(pred,y,normalize=False) # calculate number of correctly classified samples
    score = accuracy_score(pred,y,normalize=True) # calculate accuracy
    print("There are",correct,"correctly classified samples for",k+1,"-NN and",w,"voting (acc=","{:.2f}".format(score),").")
    acc_uni.append(score) # add accuracy to list for later plotting
    
    w = "distance" # distance voting
    clf = neighbors.KNeighborsClassifier(n_neighbors=k+1, weights=w) # create a Neighbours Classifier
    clf.fit(X, y) # fit the data
    pred = clf.predict(X) # predict values for all data
    correct = accuracy_score(pred,y,normalize=False) # calculate number of correctly classified samples
    score = accuracy_score(pred,y,normalize=True) # calculate accuracy
    print("There are",correct,"correctly classified samples for",k+1,"-NN and",w,"voting (acc=","{:.2f}".format(score),").\n")
    acc_dis.append(score) # add accuracy to list for later plotting
    
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1,max_k+1),acc_uni)
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.title("Uniform Voting")
plt.subplot(1, 2, 2)
plt.plot(range(1,max_k+1),acc_dis)
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.title("Distance Voting")
plt.show() 
```

## 8.2. Linearna regresija

### 8.2.1. Definicija  

**Linearna regresija** je statistička metoda koja pomaže u predviđanju veze između dve promenljive tačke podataka: njegove osobine i njegove vrednosti. Cilj je pronaći liniju (poznatu kao linija regresije) koja najbolje odgovara tačkama podataka na dijagramu sa raspršenim tačkama. Linija se zatim može koristiti za predviđanje budućih vrednosti sličnih tačaka podataka na osnovu njihovih osobina. Očigledno se koristi za rešavanje problema regresije.

U jednostavnim terminima, linearna regresija nam pomaže da razumemo odnos između dve promenljive i da na osnovu tog odnosa pravimo predviđanja. Ovaj odnos je dat modelom hipoteze:

$y_{pred}=\theta_{0}+\theta_{1}x$

Kao što vidite, model nije ništa više od jednačine prave linije u $(x,y)$ koordinatnom sistemu. Ovde je $x$ karakteristika tačke podataka, a $y_{pred}$ je predviđena vrednost tačke podataka. Na primer, $x$ može biti veličina kuće (u kvadratnim metrima), a $y_{pred}$ može biti predviđena cena te kuće. Linearna regresija nam daje model koji pokušava da pronađe vezu između veličine i vrednosti kuće. Računar to čini tako što pronalazi najbolje moguće vrednosti za **parametre** $\theta_0$ i $\theta_1$.

Ovo podešavanje parametara je zapravo deo mašinskog učenja algoritma. Računar treba skup podataka za obuku sa mnogo tačaka podataka u obliku [veličina, cena]. Koristeći te tačke podataka, algoritam pronalazi liniju koja najbolje opisuje te tačke, tako što izračunava koliko je svaka tačka udaljena od linije (što bliže, to bolje). Linija koja prilazi tačkama podataka što je moguće bliže je najbolja moguća linija (sl. 3).

Ove udaljenosti tačaka podataka od linije predstavljaju razliku između stvarne vrednosti tačke podataka $y_{act}$ i vrednosti tačke podataka koja je predviđena algoritmom (linijom) $y_{pred}$.

### 8.2.2. Funkcija troškova i optimalni parametri 

Ove stvari mogu biti vizuelno jasne i lako razumljive za nas, ali računar treba da ih prevede u matematički jezik kako bi ih interpretirao i izvršio. Dakle, kažemo da je kvalitet modela predstavljen "srednjom kvadratnom greškom (MSE) svih tačaka podataka" $J$:

$J(\theta_0,\theta_1)=\frac{1}{2m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)^2=\frac{1}{2m}\sum_{i=1}^{m}(\theta_0+\theta_1x^i-y_{act}^i)^2$

Ovde, $m$ je broj tačaka podataka u skupu za obuku. Što je ukupna greška manja, model je bolji. Dakle, algoritam zapravo pokušava pronaći parametre $\theta_0$ i $\theta_1$ koji minimiziraju funkciju $J$. Ti parametri će dati jednačinu najbolje moguće linije za dati skup podataka.

Bez ulaska u previše matematičkih detalja, algoritam postiže ovo obukom na sledeći način:


$\text{inicijalizovati } \theta_0=\theta_1=0$  
$\text{za unapred određeni broj koraka ili dok se ne dostigne konvergencija, ponoviti sledeće:}$  
$\{$  
&emsp;$\theta_0=\theta_0-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)$  
&emsp;$\theta_1=\theta_1-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)x^i$  
$\}$

U ovom slučaju, dostizanje konvergencije znači da nakon nekog vremena vrednost $J$ prestaje da se menja. Ovde, $\alpha$ je hiperparametar zvani **brzina učenja**, vrednost između 0 i 1, i određuje koliko brzo će se obuka odvijati. Male vrednosti $\alpha$ rezultiraju sporim, ali sigurnim obukama, dok prevelika vrednost može destabilizovati algoritam i proizvesti beskorisne i beskonačno velike vrednosti za parametre $\theta_0$ i $\theta_1$.

Kada je gore navedeni postupak završen, dobijamo **optimalne vrednosti** parametara i najbolju moguću liniju koja opisuje podatke. Zatim, ako nas zanima cena nove kuće, možemo ubaciti njenu veličinu (njenu karakteristiku) u naš model, videti koja će biti predviđena vrednost i uporediti je sa stvarnom vrednošću. Za evaluaciju ponovo koristimo MSE.

### 8.2.3. Primer sa jednom karakteristikom

Sledeće ćemo videti primer upotrebe linearne regresije za izgradnju modela koji predviđa nedeljni profit od turizma (u hiljadama EUR) malih gradova na Balkanu u funkciji njihovog broja stanovnika. Broj stanovnika svakog grada je karakteristika $x$, dok je profit svakog grada vrednost $y$.

Nakon uvoza, štampanja i vizualizacije podataka, podelićemo ih na skupove za obuku i testiranje i pokrenuti algoritam linearne regresije.

```python
# Import libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = pd.read_csv('https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/profit.txt', header=None, names=['Population', 'Profit']) # open data file
print ("Data preview:\n",data.head())
print ("Data statistics:\n",data.describe())

data.plot(kind='scatter', x='Population', y='Profit', figsize=(8,4));   # plot the data points
```

```python
X = data.iloc[:,0] # the features are contained in all rows, first column
y = data.iloc[:,1] # the values are contained in all rows, second column

# Split the data into traning and test subsets (ratio 9:1): 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/10, random_state=0) 
# Check the dataset and subsets dimensions: 
print('Dataset size: ', data.shape)
print('X_train size: ', X_train.shape)
print('X_test size: ', X_test.shape)
print('y_train size: ', y_train.shape)
print('y_test size: ', y_test.shape)

# Run the algorithm:
lr = LinearRegression()
lr.fit(X_train.values.reshape(-1,1),y_train) # train
# Get the coefficients: 
theta_0 = lr.intercept_ 
theta_1 = lr.coef_
print("Theta_0=","{:.2f}".format(theta_0))
print("Theta_1=","{:.2f}".format(theta_1[0]))
```

Nakon obuke algoritma na podskupu za obuku, vidimo da je jednačina za liniju:

$y_{pred}=-3.73+1.17x$

Zatim provlačimo test podatke kroz ovu jednačinu i izračunavamo srednju kvadratnu grešku kao pokazatelj uspeha modela (što je manje, to je bolje).

```python
y_pred = lr.predict(X_test.values.reshape(-1,1)) # predict
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred)) # evaluate 
```

Konačno, nacrtajmo liniju i pogledajmo kako obuhvata podatke. Takođe ćemo dodati još jedan grad i videti šta model kaže o njemu.

```python
x = np.linspace(data.Population.min(), data.Population.max(), 100)
f = theta_0 + theta_1 * x # this is the linear model for prediction with the parameters theta

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x, f, 'r') # plot the prediction line 
ax.scatter(data.Population, data.Profit) # plot the data points
ax.set_xlabel('Population (thousands)')
ax.set_ylabel('Profit (thousands)')
ax.set_title('Predicted Profit vs. Population Size')

prof = lr.predict([[20]]) # add a new town with a population of 20
print("For example, according to our model, a town with a population of 20,000 people would have a weekly profit of","{:.2f}".format(1000*prof[0]),"EUR.")
ax.scatter(20, prof, c='r'); # plot the town and it's profit value on the same figure 
```

Naravno, ovo nije savršeno prilagođavanje jer tačke podataka nisu tačno linearne. Ali prate donekle linearni trend, pa kažemo da je ovo skoro linearni problem i možemo koristiti linearnu regresiju na njemu. Linearna regresija može se koristiti samo na problemima koji su linearni ili imaju trendove koji su blizu linearnim.

### 8.2.4. Primer sa više karakteristika

U stvarnim problemima, vrednosti retko zavise samo o jednoj karakteristici. Profit grada ne zavisi samo o broju stanovnika, već i o mnogim drugim faktorima (veličina, lokacija, država, klima, itd). Slično tome, cena kuće zavisi o više stvari od njene veličine (takođe o broju soba, starosti, gradu, kvartu, itd). Zbog toga se linearna regresija vrlo retko koristi sa jednom karakteristikom, već sa više njih.

Međutim, algoritam se uopšte ne menja, bez obzira na broj karakteristika. Skup podataka može imati $n$ karakteristika i time biti n-dimenzionalan. Ne bismo mogli prikazati ili vizuelizovati podatke ako je $n>3$, ali linearna regresija će svejedno raditi. Jedina razlika je što model više neće biti dat sa 2-dimenzionalnom linijom, već sa n-dimenzionalnom hiper-ravni.

Model je sada dat sa:

$y_{pred}=\theta_{0}+\theta_{1}x_1+...+\theta_{n}x_n$  

Ovo je takođe jednačina hiper-ravni. To znači da model uopšteno ima $n+1$ parametara.

Algoritam je takođe sličan:

$\text{initialize } \theta_0=\theta_1=...=\theta_n=0$  
$\text{for a predetermined number of steps or until convergence reached, repeat the following:}$  
$\{$  
&emsp;$\theta_0=\theta_0-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)$  
&emsp;$\theta_1=\theta_1-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)x_1^i$  
&emsp;...  
&emsp;$\theta_n=\theta_n-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)x_n^i$  
$\}$  

Međutim, postoji jedna značajna razlika: prvo moramo standardizovati podatke. Standardizacija znači da moramo oduzeti srednju vrednost podataka od svake tačke podataka, a zatim rezultat podeliti sa standardnom devijacijom podataka. Kao rezultat toga, dobijamo nove podatke koji sadrže iste informacije kao i ranije, ali sada imaju srednju vrednost 0 i standardnu devijaciju 1. Ovo mora biti urađeno kada radimo sa više karakteristika, jer su te karakteristike uopšteno različito raspoređene u prostoru stanja. Na primer, karakteristika koja označava veličinu kuće može se kretati od 800 do 4500 kvadratnih metara, ali karakteristika koja označava broj soba kreće se samo od 1 do 5. Da bismo izbegli ovu nejednakost koja može prouzrokovati greške i usporiti izvršavanje algoritma, moramo standardizovati.


```python
# Import libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = pd.read_csv('https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/prices.txt', header=None, names=['Size', 'Bedrooms', 'Price']) # open data file
print ("Data preview:\n",data.head())
print ("Data statistics:\n",data.describe())

# Standardize the features in the multiple variable regression: 
data_s = (data - data.mean() ) / data.std()
print ("Standardized data:")
print (data_s.head())
print (data_s.describe())

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data_s.Size, data_s.Bedrooms, data_s.Price)  # plot the data points
```

```python
X = data_s.iloc[:,:2] # the features are contained in all rows, first two columns
y = data_s.iloc[:,2] # the values are contained in all rows, third column

# Split the data into traning and test subsets (ratio 9:1): 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/10, random_state=0) 
# Check the dataset and subsets dimensions: 
print('Dataset size: ', data.shape)
print('X_train size: ', X_train.shape)
print('X_test size: ', X_test.shape)
print('y_train size: ', y_train.shape)
print('y_test size: ', y_test.shape)

# Run the algorithm:
lr = LinearRegression()
lr.fit(X_train,y_train) # train
# Get the coefficients: 
theta_0 = lr.intercept_ 
theta_1 = lr.coef_[0]
theta_2 = lr.coef_[1]
print("Theta_0=","{:.2f}".format(theta_0))
print("Theta_1=","{:.2f}".format(theta_1))
print("Theta_2=","{:.2f}".format(theta_2))
```

Nakon što smo obučili algoritam na trening podskupu, vidimo da je jednačina za 2-dimenzionalnu ravan:

$y_{pred}=-0.07+0.81x_1-0.03x_2$

Zatim pokrećemo testne podatke kroz ovu jednačinu i izračunavamo srednje kvadratno odstupanje kao indikaciju uspešnosti modela (što je niže, to je bolje).

```python
y_pred = lr.predict(X_test) # predict
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred)) # evaluate 
```

Na kraju, crtajmo ravan i pogledajmo kako obuhvata podatke. Takođe ćemo dodati dodatnu kuću i videti šta model govori o njoj.

```python
x12 = np.linspace(data_s.Size.min(), data_s.Size.max(), 100)
x22 = np.linspace(data_s.Bedrooms.min(), data_s.Bedrooms.max(), 100)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data_s.Size, data_s.Bedrooms, data_s.Price, c='c')  # plot the data points
X12,X22 = np.meshgrid(x12,x22)

f = theta_0 + theta_1 * X12 + theta_2 * X22 # this is the linear model for prediction with the parameters theta

ax.plot_surface(X12, X22, f) # plot the prediction plane
ax.set_xlabel('Size (sq feet) - standardized')
ax.set_ylabel('Bedrooms - standardized')
ax.set_zlabel('Price [EUR] - standardized')
ax.set_title('Predicted Price vs. House Size')

new_house = data.iloc[-1,:] # take a new house (the last data point from the file for example: 1203 sq feet, 3 bedrooms)
new_house = (new_house - data.mean() ) / data.std() # standardize it
new_X = new_house[:2] # take only the features
price = lr.predict(new_X.values.reshape(1,-1)) # predict the value of this data point 
ax.scatter(new_X[0], new_X[1], price, c='r'); # plot the house and it's price on the same figure in yellow
```

Tačka se tačno nalazi na ravni, ali se blago razlikuje od svoje stvarne vrednosti u blizini.

Ponovo je očigledno da ravan ne pokriva sve tačke savršeno, što je nemoguće jer nisu linearni. Ali uspešno hvata trend podataka koji je blizu linearnog. To je najviše što možemo tražiti od linearne regresije kada je koristimo za rešavanje stvarnih, nelinearnih problema.

## 8.3. Logistička regresija
### 8.3.1. Definicija

**Logistička regresija** je statistička metoda koja se koristi u mašinskom učenju za predviđanje binarnog ishoda (tj. dva moguća rezultata, kao što su da/ne ili tačno/netačno) na osnovu jedne ili više ulaznih promenljivih. Kao što smo prethodno razgovarali, ovo je zapravo proces binarne klasifikacije.

Zamislite da pokušavate da predvidite da li će student položiti ili pasti na ispitu na osnovu broja sati koje je proveo učeći. Ove informacije možete predstaviti na grafikonu, sa brojem sati učenja na X-osi i prolaznim/neuspešnim ishodom na Y-osi.
<figure>
<img src="https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/logreg.jpeg" style="height: 400px;"/> 
<figcaption align = "center"> Slika 4 (preko Wikimedia Commons) </figcaption>
</figure>

Na grafikonu iznad, tačke predstavljaju pojedinačne studente i njihove sate učenja i rezultate ispita. Primetite da kako broj sati učenja raste, verovatnoća prolaska ispita takođe raste. Ovde logistička regresija dolazi u igru - pomaže nam da pronađemo najbolju liniju koja razdvaja dva ishoda što je moguće bolje. Primetite da je linija sada zakrivljena, za razliku od prave linije koju smo imali sa linearnom regresijom.

Linija se može matematički predstaviti jednačinom, koja uzima ulaznu promenljivu (u ovom slučaju, broj sati učenja) i izlaznu verovatnoću prolaska ispita. Na primer, jednačina za liniju na grafikonu iznad može izgledati ovako:

$y_{pred}=\frac{1}{1+e^{-(\theta_0+\theta_1x)}}$

Ovo je model ili hipoteza. Ovde je $x$ sati učenja (karakteristika), a $\theta_0$ i $\theta_1$ su ponovo parametri koje algoritam treba izračunati za najbolju moguću liniju. Konačno, $y_{pred}$ je vrednost koju će ovaj model predvideti - primetite da se kreće između 0 (ako je eksponent $e$ u zagradama vrlo negativan, imenilac je pozitivno beskonačan) i 1 (ako je eksponent $e$ u zagradama vrlo pozitivan, imenilac je 1). Dakle, model u logističkoj regresiji izlazi sa vrednošću verovatnoće između 0 i 1 (0% i 100%) koliko je verovatno da se ostvari ishod problema. Ali pošto je ovo binarna klasifikacija, kažemo da je sredina ovog opsega 0.5 (50%) prag kada odlučujemo:

* ako $y_{pred}>=0.5$, onda model predviđa 1 ili TAČNO;  
* ako $y_{pred}<0.5$, onda model predviđa 0 ili NETAČNO.  

Ovaj pristup se naziva logistička regresija iako se koristi samo za probleme klasifikacije. Razlog je što je model ponovo regresijski model (izlazi sa kontinuiranom vrednošću $y_{pred}$) kako se može videti na slici 4. Ali upravo je uvođenje praga koji smo gore definisali čini ovaj pristup klasifikacijskim pristupom (izlazi 0 ili 1). Ime "logističko" potiče od vrste funkcije koja se koristi u modelu, koja se naziva logistička funkcija.

Kao i kod linearne regresije, logistička regresija se takođe može koristiti u problemima sa više karakteristika, što je dobro jer većina stvarnih problema ima mnogo više od jedne karakteristike. U slučaju $n$ karakteristika (n-dimenzionalni problem), model je:

$y_{pred}=\frac{1}{1+e^{-(\theta_0+\theta_1x_1+...+\theta_nx_n)}}$

Bez ulaska u previše matematičkih detalja, računar optimizuje parametre $\theta$ na sličan način kao što je prikazano u linearnoj regresiji.

Na kraju, način na koji procenjujemo uspeh modela logističke regresije sličan je onome koji smo prethodno predstavili u k-NN. Brojimo sve tačno klasifikovane primere iz test podskupa podataka, i taj broj delimo sa ukupnim brojem podataka u test podskupu, čime dobijamo metriku tačnosti.

$\text{PRECIZNOST}=\frac{\text{BROJ TAČNO KLASIFIKOVANIH PODATAKA}}{\text{UKUPAN BROJ PODATAKA}}$  

### 8.3.2. Primer klasifikacije sa više promenljivih

Pogledajmo primer gde imamo podatke o 100 studenata koji su se prijavili na Tehnički univerzitet. Znamo njihove rezultate na dva prijemna ispita iz matematike i fizike (to su osobine $x_1$ i $x_2$) i znamo da li su primljeni (1) ili nisu primljeni (0) na univerzitet (to je klasa). Obučimo klasifikator logističke regresije koji gradi model i može predvideti da li se budući studenti mogu upisati na osnovu rezultata ispita.

```python
# Import the libraries and the data:
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix

data = pd.read_csv('https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/exams.txt', header=None, names=['MathScore', 'PhysicsScore', 'Accepted?'])
print("All data:\n",data)
X = data.drop(columns=['Accepted?']) # x is all columns but 'Accepted?'
y = data['Accepted?'] # y is the 'Accepted?' column 
print("X:\n",X)
print("y:\n",y)

# Plot the data
positive = data[data['Accepted?'].isin([1])]  
negative = data[data['Accepted?'].isin([0])]
fig, ax = plt.subplots(figsize=(8,4))  
ax.scatter(positive['MathScore'], positive['PhysicsScore'], s=50, c='b', marker='o', label='Accepted')  
ax.scatter(negative['MathScore'], negative['PhysicsScore'], s=50, c='r', marker='x', label='Not Accepted')  
ax.legend()  
ax.set_xlabel('Math Score')  
ax.set_ylabel('Physics Score');
```

Vidimo da su podaci prilično očekivani - oni sa višim rezultatima se uglavnom upisuju na univerzitet. Ali hajde da vidimo gde je tačno razdvajanje. Sada obučavamo model logističke regresije.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/10, random_state=0) # get train and test subsets
model = LogisticRegression() # create the classifier
model.fit(X_train, y_train) # train the classifier
# Return the coefficients theta: 
print("Theta_0 = ","{:.2f}".format(model.intercept_[0]))
print("Theta_1 = ","{:.2f}".format(model.coef_[0,0]))
print("Theta_2 = ","{:.2f}".format(model.coef_[0,1]))
```

Nakon obučavanja algoritma na trening podskupu, vidimo da je model:

$y_{pred}=\frac{1}{1+e^{-(-24.14+0.19x_1+0.20x_2)}}=\frac{1}{1+e^{(24.14-0.19x_1-0.20x_2)}}$ 

Zatim testiramo podatke kroz ovu jednačinu kako bismo dobili predviđanja i izračunali tačnost u poređenju sa stvarnim vrednostima.

```python
y_pred = model.predict(X_test)
acc = metrics.accuracy_score(y_test,y_pred)
print("Accuracy=","{:.2f}".format(acc))
```

Hajde da rezultate prikažemo na drugačiji način. Tačnost daje samo jednu vrednost za kvantifikaciju modela, ali ako koristimo matricu konfuzije, možemo dobiti više informacija. Matrica konfuzije ima 2 reda i 2 kolone. Redovi predstavljaju stvarne vrednosti klasa (0 u redu 0 i 1 u redu 1). Kolone predstavljaju predviđene vrednosti klasa (0 u redu 0 i 1 u redu 1).

* Broj na poziciji (0,0) predstavlja **tačne negativne TN**. To je broj podataka koji su iz klase 0 i tačno su predviđeni kao klasa 0.
* Broj na poziciji (0,1) predstavlja **lažne pozitive FP**. To je broj podataka koji su iz klase 0 i netačno su predviđeni kao klasa 1.
* Broj na poziciji (1,0) predstavlja **lažne negativne FN**. To je broj podataka koji su iz klase 1 i netačno su predviđeni kao klasa 0.
* Broj na poziciji (1,1) predstavlja **tačne pozitive TP**. To je broj podataka koji su iz klase 1 i tačno su predviđeni kao klasa 1.

```python
plot_confusion_matrix(model,X_test,y_test,cmap='Reds');
```

Kao što vidimo, u test podskupu bilo je ukupno 10 podataka, 5 pozitivnih i 5 negativnih. Sva 5 pozitivna su tačno klasifikovana (vidi desnu kolonu), ali je bilo jedne greške pri klasifikaciji 5 negativnih (vidi levu kolonu). To potvrđuje prethodno dobijeni rezultat tačnosti od 90%. Matrica konfuzije je koristan alat pri vizualizaciji tačnosti i mnogo informativnija od same vrednosti tačnosti. Često se koristi sa svim algoritmima klasifikacije.

#### 8.3.2.1. 8.3.2.1. Linija razdvajanja

Na kraju, postoji još jedna važna stvar koju treba napomenuti. Iako model logističke regresije više ne daje pravu liniju, unutar njega se krije prava linija (ili hiper-ravan u opštem slučaju). Eksponent $e$ unutar logističke funkcije je zapravo jednačina za liniju ili ravan koja razdvaja podatke u prostoru osobina. Da bismo to pokazali, hajde da nacrtamo liniju formiranu parametrima koje smo dobili u prostoru osobina.

$x_2 = -\frac{\theta_0}{\theta_2}-\frac{\theta_1}{\theta_2}x_1=\frac{24.14}{0.20}-\frac{0.19}{0.20}x_1$ 

```python
x = np.linspace(data.MathScore.min(), data.MathScore.max(), 100)
f = -model.intercept_[0]/model.coef_[0,1] - model.coef_[0,0]*x/model.coef_[0,1]
fig, ax = plt.subplots(figsize=(8,4))  
ax.scatter(positive['MathScore'], positive['PhysicsScore'], s=50, c='b', marker='o', label='Accepted')  
ax.scatter(negative['MathScore'], negative['PhysicsScore'], s=50, c='r', marker='x', label='Not Accepted')  
ax.legend()  
ax.set_xlabel('Math Score')  
ax.set_ylabel('Physics Score');
ax.plot(x, f, 'g', label='Decision Boundary');
```

Očigledno se neke tačke podataka nalaze na pogrešnoj strani linije, ali to je očekivano jer se podaci ne mogu linearno razdvojiti. Ova linija razdvajanja (takođe nazvana granica odlučivanja) će uvek biti prava, pa će algoritam praviti greške. Ali ako je granica gotovo linearna sa nekim izuzecima, onda će logistička regresija raditi sasvim dobro (primetite sličnosti sa linearnom regresijom u ovom smislu).

## 8.4. Stabla odlučivanja (Decision trees)

### 8.4.1. Definicija

Stablo odlučivanja je vrsta algoritma mašinskog učenja koji se može koristiti za rešavanje problema tako što pravi predikcije na osnovu podataka. To funkcioniše tako što razbija složeni problem na manje, lakše rešive delove, a zatim donosi odluke na osnovu podataka dostupnih na svakom koraku. Ne gradi eksplicitan model za sve podatke unapred, i u tom smislu se razlikuje od logističke i linearne regresije. Može se koristiti i za klasifikaciju i za regresiju, ali je lakše razumeti i predstaviti u kontekstu klasifikacije, pa ćemo to uraditi ovde.

Zamislite da želite da predvidite vrstu voća koje imate na osnovu boje i oblika. Možete početi tako što postavite niz pitanja, kao što su:

        * Da: Voće je jagoda.

* b. Ne:
    * i. Da li je voće crveno?

Svaki korak u procesu pomaže da se suze moguće opcije, dok se ne donese konačna predikcija. Ovo je slično kako stablo odlučivanja funkcioniše u mašinskom učenju, gde je svaki korak predstavljen **čvorom** u stablu, a odluke donete u svakom čvoru vode do različitih **grana**. Konačna predikcija se pravi na **listu čvora**. Tako se stablo odlučivanja može vizualizovati kao dijagram u obliku stabla, gde grane predstavljaju različite odluke donete u svakom čvoru, a listovi predstavljaju konačne predikcije.

U ovom primeru, prva odluka je da li je voće žuto ili ne. Ako je žuto, algoritam proverava da li je dugo i tanko ili ne. Ako voće nije žuto, algoritam proverava da li je narandžasto ili ne. Ako jeste, voće se klasifikuje kao pomorandža ako je manje od grejpfruta ili kao grejpfrut ako nije manje. Ako nije narandžaste boje, voće se klasifikuje kao jabuka.

Kao što možete videti, kretanje kroz stablo odlučivanja je vrlo jednostavno, jer sve što radimo je odgovaranje na pitanja i biranje grana u zavisnosti od naših odgovora, dok ne stignemo do lista. Ovo je tačno kako izgleda testna faza, kada se tačke podataka iz testnog podskupa ubacuju u stablo i "prelaze ga" u zavisnosti od vrednosti njihovih karakteristika, dok ne dođu do lista koji sadrži njihovu klasu/vrednost. Ali, teško pitanje je: kako se stablo gradi koristeći podatke za obuku, tj. kako se pitanja u čvorovima biraju?

### 8.4.2. Nečistoća i izgradnja stabla

Stabla odlučivanja koriste metriku zvanu Ginijeva nečistoća kako bi odlučili o tome kako da podele podatke, odnosno na osnovu kojih osobina i vrednosti da granaju čvorove. Ginijeva nečistoća je mera koliko je grupa stvari "pomešana". Zamislite da imate činiju sa M&M bombonama i želite da ih podelite u dve zasebne činije po boji. Cilj je da imate jednu činiju sa svim plavim bombonama i drugu činiju sa svim crvenim bombonama. Što su bombone više pomešane, Ginijev indeks nečistoće biće veći. Ako su dve činije savršeno podeljene po boji, Ginijev indeks nečistoće biće 0, što ukazuje na to da bombone uopšte nisu pomešane. U mašinskom učenju koristimo Ginijevu nečistoću kako bismo procenili koliko dobro se skup stavki može podeliti u različite grupe na osnovu određenih karakteristika. Cilj je imati nisku Ginijevu nečistoću, što znači da su stavke dobro podeljene u grupe sa sličnim karakteristikama.

Prilikom treniranja stabla, računar isprobava svaku osobinu kao koreni čvor. Grananje vrši i izračunava nečistoću nakon podele. Osobina koja daje najniži skor Ginijeve nečistoće bira se kao koren i grananje se vrši. Zatim se postupak ponavlja ponovo u svim čvorovima, na svakoj dubini i za sve karakteristike. Ovo se nastavlja dok ne dostignemo unapred određenu maksimalnu dubinu ili dok nemamo više čvorova za grananje (tj. dok ne dođemo do svih listova čvorova). Dostizanje lista čvora znači da imamo Ginijevu nečistoću od 0, jer su svi podaci jasno i tačno klasifikovani u tom čvoru.

Ovaj pristup vam može izgledati dug i naporan - isprobavanje svih karakteristika kod svakog čvora zvuči zamorno! Ali imajte na umu da, što dublje idemo u stablo, manje podataka obrađujemo, jer svaki čvor analizira samo podatke koji zadovoljavaju grane koje vode do njega. Stoga stvari idu brže i računar može priuštiti da isprobava različite karakteristike iznova i iznova. 

Isprobajmo nekoliko različitih stabala odlučivanja sa iris skupom podataka kako bismo ilustrovali sve što smo do sada rekli. Kao i ranije, uvešćemo podatke, podeliti ih na trening i test podskupove, a zatim trenirati stabla odlučivanja pre nego što testiramo njihovu tačnost. Isprobaćemo 3 različita stabla odlučivanja i uporediti njihove tačnosti: jedno sa maksimalnom dubinom od 2, jedno sa maksimalnom dubinom od 5 i jedno bez zadate maksimalne dubine (tj. grananje će se vršiti dok ne dostignemo sve listove čvorova). 

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the Iris dataset:
iris = load_iris()

# Create a DataFrame with the features and target:
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
print(df)

# Split the data into training and testing sets:
X_train,X_test,y_train,y_test = train_test_split(df[iris['feature_names']],df['target'],test_size=0.25,random_state=0)

# Fit a decision tree model with a maximum depth of 2:
clf1 = DecisionTreeClassifier(max_depth=2, random_state=0)
clf1.fit(X_train, y_train)
# Evaluate the accuracy of the model on the test data:
y_pred1 = clf1.predict(X_test)
print("Accuracy with max_depth=2:", accuracy_score(y_test, y_pred1))

# Fit a decision tree model with a maximum depth of 5:
clf2 = DecisionTreeClassifier(max_depth=5, random_state=0)
clf2.fit(X_train, y_train)
# Evaluate the accuracy of the model on the test data
y_pred2 = clf2.predict(X_test)
print("Accuracy with max_depth=5:", accuracy_score(y_test, y_pred2))

# Fit a decision tree model with a maximum depth of None (no limit on the depth):
clf3 = DecisionTreeClassifier(max_depth=None, random_state=0)
clf3.fit(X_train, y_train)
# Evaluate the accuracy of the model on the test data:
y_pred3 = clf3.predict(X_test)
print("Accuracy with max_depth=None:", accuracy_score(y_test, y_pred3)) 
```

Vidimo da je najbolja tačnost postignuta sa maksimalnom dubinom od 5 (ili možda čak i pre), ali maksimalna dubina od 2 nije dovoljna. Kako bismo to ilustrovali, nacrtajmo stablo odlučivanja sa maksimalnom dubinom i pogledajmo koliko duboko ide.

```python
import matplotlib.pyplot as plt
# Plot the tree with maximum depth (clf3):
plt.figure(figsize=(20, 20))
tree.plot_tree(clf3, filled=True, feature_names=df.columns)
plt.show()
# Left -> True; Right >- False
```

Dakle, stablo prvo deli koren na osnovu karakteristike širine latice, postavljajući prag odluke na 0,8 cm. Cvetovi sa manjom širinom latice odmah se klasifikuju kao klasa 1 iris jer nailazimo na podskup trening podataka koji ne sadrži druge cvetove (ima Ginijevu nečistoću od 0). Zato je ovo list. Ostatak podataka zatim se deli na osnovu karakteristike dužine latice, sa pragom od 4,95 cm, i tako dalje. Kao što možemo videti, stablo ima dubinu od 4 pre nego što dođe do svih listova čvorova sa Ginijevom nečistoćom od 0, pa je to maksimalna dubina.

Ovo je jedna od velikih prednosti stabala odlučivanja - možemo jasno videti i razumeti zašto klasifikuju podatke na način na koji to rade. Sve što treba da uradimo je da nacrtamo stablo i pratimo čvorove i grane za svaki uzorak. Zato se stabla odlučivanja nazivaju "white-box model" (kao u "transparentna kutija").

Takođe možemo videti da se Ginijeva nečistoća smanjuje kako idemo dublje i stvari postaju jasnije - počinjemo sa nečistoćom od 0,665 u korenu i završavamo sa nečistoćom od 0 na svim listovima. Takođe, primetite kako počinjemo sa ukupno 112 uzorka (podaci za trening čine 3/4 originalnog skupa podataka od 150 cvetova), ali se broj uzoraka smanjuje kako idemo dublje u stablo.

Sledeće, pogledajmo matricu konfuzije za test podskup na ovom stablu, ali i na stablu sa maksimalnom dubinom od 2, koje je imalo niži rezultat tačnosti.

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf3,X_test,y_test,cmap='Reds');
plot_confusion_matrix(clf1,X_test,y_test,cmap='Reds');
```

Od 38 uzoraka koji su korišćeni za testiranje stabla sa maksimalnom dubinom, samo je jedan pogrešno klasifikovan (predviđena klasa 2, ali zapravo klasa 1). Stablo sa maksimalnom dubinom od 2 napravilo je 3 greške više.

Imajte na umu da povećanje maksimalne dubine stabla generalno dovodi do boljih performansi na podacima za obuku, ali može dovesti do **preprilagođavanja** (tj. model je previše slošežen i ne generalizuje dobro na nove podatke). Dobra praksa je podešavanje maksimalne dubine kako bi se pronašla najbolja ravnoteža između tačnosti i generalizacije.

### 8.4.3. Veći primer

Hajde da probamo primer sa većim skupom podataka, sa više osobina i uzoraka. Sadrži informacije o zdravlju i fiziologiji brojnih pacijenata (osobine) i informacije o tome da li su im dijagnostikovani dijabetes ili ne (klasa 1 ili 0). Ako obučimo dobar klasifikacioni model, možemo ga koristiti za predviđanje ili barem pomoć u dijagnostikovanju dijabetesa kod novih pacijenata u budućnosti.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the dataset:
data = pd.read_csv("https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/diabetes.csv")

# Split the data into features and target:
X = data.drop("Outcome", axis=1)
y = data["Outcome"]

# Split the data into training and testing sets:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train a decision tree classifier:
dtree = DecisionTreeClassifier(max_depth=None, random_state=0)
dtree.fit(X_train, y_train)

# Predict the outcome of test data:
y_pred = dtree.predict(X_test)

# Calculate the accuracy of the model:
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ","{:.2f}".format(accuracy))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(20, 20))
tree.plot_tree(dtree, filled=True, feature_names=X.columns)
plt.show()
# Left -> True; Right >- False
```

Stablo je sada mnogo veće i dostiže maksimalnu dubinu od 16. Hajde da vidimo kako se tačnost testiranja menja na različitim nivoima dubine.

```python
acc_train = [] # list for all the training accuracies
acc_test = [] # list for all the testing accuracies

for i in range(16): # try depths from 1 to 16
    dtree = DecisionTreeClassifier(max_depth=i+1, random_state=0) # generate tree with appropriate depth
    dtree.fit(X_train, y_train) # train tree
    y_pred_train = dtree.predict(X_train) # test tree on training data
    y_pred_test = dtree.predict(X_test) # test tree on testing data
    accuracy_train = accuracy_score(y_train, y_pred_train) # calculate training accuracy
    accuracy_test = accuracy_score(y_test, y_pred_test) # calculate testing accuracy
    acc_train.append(accuracy_train) 
    acc_test.append(accuracy_test) 

# Plot accuracies for different depths:
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1,17),acc_train)
plt.xlabel("Maximum depth")
plt.ylabel("Accuracy")
plt.title("Tree training accuracy depending on maximum depth")
plt.subplot(1, 2, 2)
plt.plot(range(1,17),acc_test)
plt.xlabel("Maximum depth")
plt.ylabel("Accuracy")
plt.title("Tree testing accuracy depending on maximum depth")
plt.show() 
```

Kako se očekivalo, ako testiramo stablo na podacima za obuku (koje je već videlo tokom obuke), tačnost se poboljšava sa dubinom i čak dostiže 100%.

Ali zanimljivo je videti da se pri testiranju na neviđenim podacima najbolja tačnost ne javlja za najdublje stablo. Umesto toga, javlja se za dubinu od 10. Razlog za to je što dublja stabla imaju tendenciju da preprilagode podatke, tj. "previše razmišljaju" kada traže vezu između osobina i vrednosti/klase. Realni ekvivalent ovome bio bi student koji uči pitanja i odgovore napamet; ako mu postave ta pitanja, znaće odgovore savršeno. Ali ako mu postave bilo koje drugo pitanje vezano za materijal, ali drugačije od onog koje su naučili, suočavaće se sa poteškoćama jer su naučili pitanja i odgovore bez razumevanja.

Kao što smo rekli ranije, želimo dobru ravnotežu između tačnosti i generalizacije, pa bi trebalo proveriti tačnost za različite vrednosti hiperparametara (što u ovom slučaju predstavlja maksimalna dubina stabla) pre nego što se odlučimo za najbolji.
