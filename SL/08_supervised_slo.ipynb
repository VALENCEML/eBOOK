### Osnove

Nadzorovano učenje je vrsta strojnega učenja, pri kateri računalnik dobi označene primere podatkov in želeni rezultat, na podlagi teh informacij pa se nauči, kako ustvariti pravilen rezultat, ko dobi nove podatke. V bistvu je to podobno, kot če bi učencu pokazali, kako naj reši matematično nalogo, nato pa bi vadil podobne naloge, da bi se izboljšal pri njihovem reševanju. Pri nadzorovanem učenju računalnik deluje kot učenec, označeni podatki pa so kot primeri, ki so učencu na voljo. Cilj je, da računalnik na podlagi vhodnih podatkov pripravi natančne napovedi ali odločitve, podobno kot bi moral biti učenec sposoben uporabiti naučeno snov pri reševanju novih problemov. Koncept nadzorovanega učenja je na preprost način prikazan na sliki 1, kjer algoritem poišče zeleno črto, ki ločuje podatke, in jih tako razvrsti glede na njihovo vrednost (X ali O).

### Značilnosti atributov in atributni prostor 

Algoritmi strojnega učenja gradijo te modele tako, da preučujejo **atribute** podatkov za učenje. Atributi so značilnosti podatkov, ki se lahko uporabijo za njihov opis. Če bi na primer poskušali razvrstiti različne živali, bi lahko bile nekatere značilnosti njihova teža, višina, barva, število nog itd. Če bi skušali opisati hišo, bi bile nekatere značilnosti lahko velikost, starost, število sob, število nadstropij hiše, itd.  

Zelo pomembno je opozoriti, da so atributi vnaprej določene vrednosti, ki so priložene podatkom. Algoritmi strojnega učenja nimajo vpliva na atribute; nasprotno, pri oblikovanju modela morajo upoštevatiatribute podatkov za učenje, pri testiranju modela pa atribute testnih podatkov.   

**Atributni prostor** je prostor, ki ga določajo atributi. Predstavljamo si ga lahko kot graf, kjer vsaka os predstavlja eno značilnost, tako da če imajo podatki _n_ značilnosti, jih lahko predstavimo z n-razsežnim koordinatnim sistemom. Prostor lastnosti je sestavljen iz vseh možnih točk v tem grafu, ki jih je mogoče ustvariti z združevanjem vrednosti za vsak atribut. Z drugimi besedami, prostor atribut so vse možne kombinacije vrednosti značilk, ki opisujejo podatkovno točko.  
### Klasifikacija in regresija  

Dve najpogostejši vrsti težav, ki se lahko pojavita pri nadzorovanem učenju, sta klasifikacija in regresija, odvisno od narave vrednosti podatkov.  

* **Klasifikacija** se uporablja za napovedovanje **kategorične oznake** ali **razreda**, na primer "pes" ali "mačka". Računalniku so dani primeri podatkov za učenje z ustrezno oznako razreda, računalnik pa se na podlagi teh informacij nauči, kako nove podatke razvrstiti v pravilen razred. Računalniku lahko na primer pokažemo slike različnih živali in mu povemo, v katero kategorijo spada vsaka od njih (npr. "pes", "mačka", "ptica" itd.). Ko računalnik dobi novo sliko živali, jo mora znati razvrstiti v pravilno kategorijo.  

* **Regresija** se uporablja za napovedovanje **kontinuirane vrednosti**, na primer cene hiše. Računalnik dobi primere učnih podatkov z ustrezno zvezno vrednostjo in se na podlagi teh informacij nauči napovedovati nove zvezne vrednosti. Na primer, računalniku lahko prikažemo velikost in lokacijo različnih hiš ter mu povemo njihove cene. Ko dobi podatke o velikosti in lokaciji nove hiše, bi moral biti sposoben napovedati njeno ceno.

Če povzamemo, se klasifikacija uporablja za napovedovanje oznak razredov, regresija pa za napovedovanje zveznih vrednosti.  

V tem poglavju bomo preučili več metod za klasifikacijo in regresijo.  

## 8.1. k-NN

### 8.1.1. Definicija

Metoda **K-najbližjih sosedov (k-NN)** je algoritem strojnega učenja, ki se uporablja za razvrščanje. Je eden najpreprostejših pristopov, ki se pogosto uporablja kot primerjalno merilo. 

Predstavljajte si, da imate nabor označenih podatkovnih točk, ki so slike dveh različnih vrst/kategorij živali (mačke in psi). k-NN deluje tako, da poišče _k_ najbližjih podatkovnih točk novi podatkovni točki in nato novo podatkovno točko uvrsti v kategorijo, ki je med "k" najbližjimi točkami najpogostejša. Pri tem je _k_ število, ki se imenuje tudi **hiperparameter** in ga morate izbrati, preden zaženete algoritem.

Če na primer izberete k=3 in imate novo podatkovno točko (sliko), ki jo želite razvrstiti kot mačko ali psa, bo k-NN našel 3 najbližje (tj. najbolj podobne) podatkovne točke tej novi točki v označenih podatkih. Če sta dve od teh treh najbližjih točk označeni kot "mačka" in ena kot "pes", bo k-NNN novo podatkovno točko razvrstil kot "mačko". To pomeni, da k-NN pri razvrščanju novih podatkov uporablja idejo "zmage večine" med najbližjimi podatkovnimi točkami. Zato velja za "leni" učni algoritem, saj nima faze učenja in za napovedovanje preprosto uporablja označene podatke. Za razliko od algoritmov, ki bodo prikazani pozneje, k-NN ne gradi modela ali hipoteze in vsako novo podatkovno točko takoj razvrsti.  

Algoritem določi podobnost med dvema podatkovnima točkama tako, da izračuna **razdaljo** med njima v prostoru značilnosti. Razdaljo med dvema točkama _X_ in _Y_ v n-razsežnem prostoru značilnosti lahko izračunamo kot:  

$D(X,Y)=\sqrt{\sum_{i=1}^n |x_{(i)}-y_{(i)}|^{2}}$

Pri "glasovanju", ki določa večino sosedov, obstajata dve možnosti:  
* **enakomeren** pristop daje enako pomembnost vsem _k_ sosedom nove podatkovne točke. Glas vsakega soseda ima vrednost 1.  
* Pristop **oddaljenosti** daje _k_ sosedom različno pomembnost glede na njihovo oddaljenost od nove podatkovne točke. Pri glasovanju ima vsak sosedov glas vrednost $\frac{1}{D}$. To pomeni, da so bližji sosedje pomembnejši in vplivnejši od tistih, ki so bolj oddaljeni.

### 8.1.2. Primer z naborom podatkov Iris  

Sedaj pokažimo uporabo k-NN za razvrščanje nekaterih cvetov iz rodu Iris. To je dobro znan nabor podatkov iz leta 1936, sestavljen iz 150 vzorcev cvetov irisa treh različnih vrst: setosa, virginica in versicolor (po 50 od vsake). To so razredi vzorcev. Za vsak cvet so na voljo tudi 4 druge znane vrednosti: dolžina in širina okrasnega lista, dolžina in širina cvetnega lista. To so atributi, kar pomeni, da imamo štiridimenzionalni problem. Čeprav je ta nabor podatkov star in precej majhen, se pogosto uporablja kot preprosto orodje za demonstracijo algoritmov za razvrščanje, kot je k-NN.  

V nadaljevanju uvozimo nabor podatkov, vzamemo samo prva dva atributa (dolžina in širina čašnega lista), da lahko podatke vizualiziramo v dvodimenzionalnem prostoru značilnosti, in preizkusimo algoritem k-NN za razvrščanje cvetov v enega od treh možnih razredov (vrst).

```python
# vključitev knjižnic in iris podatkovne zbirke
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from sklearn.metrics import accuracy_score
iris = datasets.load_iris()

X = iris.data[:, :2] # vzami samo prva 2 atributa
y = iris.target
print("Na voljo je",X.shape[0],"podatkovnih točk. Vsaka ima",X.shape[1],"atributov.")

# ustvari barve za vse tri razrede
cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

n_neighbors = 15 # nastavi število sosedov k
acc = [] # ustvari seznam, ki bo shranil točnosti

for weights in ['uniform', 'distance']: # uporabi uniformno glasovanje ali glasovanje o razdalji
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights) # ustvari k-NN klasifikator
    clf.fit(X, y) # nauči model
    pred = clf.predict(X) # napovej vrednosti za vse podatke
    acc.append(accuracy_score(pred,y)) # primerjaj dejanske podatke ter napovedane podatke
    print(accuracy_score(pred,y,normalize=False),"pravilnih primerov je doseženih z",weights,"glasovanjem.")

#### Naslednji del je kompliciran in se uporablja za slikanje. Ni ga potrebno analizirati. ####
    # Plot the results. Assign a color to each point in the coordinate system:
    # Nariši rezultate. Dodeli barvo vsaki točki v koordinatnem sistemu
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # doseg na x-osi
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # doseg na y-osi
    h = .02  # step size in the mesh
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))    
   
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]) # napoved vrednosti vsaki točki na grafu
    
    # Premakni rezultate na barvni graf
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Nariši točke, ki se uporabljajo za učenje modela
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xlabel("Dolžina lista")
    plt.ylabel("Širina lista")
    plt.title("3-razredna klasifikacija (k = %i, metoda = '%s')"
              % (n_neighbors, weights))
plt.show()
#### Komplicirani del se zaključi tukaj. ####

print("Točnost dosežena z uniform glasovanjem je","{:.2f}".format(acc[0]),".\n")
print("Točnost dosežena z glasovanjem z razdaljo (angl.distance) je","{:.2f}".format(acc[1]),".\n")
```

**Točnost** je metrika, ki se uporablja pri evalvaciji in ocenjuje, kako dobro je algoritem napovedal podatke. Prešteje število uspešno razvrščenih podatkovnih točk in ga deli s skupnim številom podatkovnih točk:  

$\text{TOČNOST}=\frac{\text{ŠTEVILO PRAVILNO NAPOVEDANIH PODATKOVNIH TOČK}}{\text{ŠTEVILO VSEH PODATKOVNIH TOČK}}$

Ima vrednost med 0 (ni bilo pravilno napovedano) in 1 (vse je bilo pravilno napovedano). 

Kot je razvidno, je točnost nekoliko boljša pri uporabi glasovanja z razdaljo. To ni vedno pravilo in rezultati se razlikujejo od primera do primera, vendar na splošno uporaba glasovanjaz radaljo prispeva k boljšemu in zanesljivejšemu klasifikatorju, ki daje prednost neposredni soseščini podatkovne točke.  

Zgornjo kodo poskusite zagnati tudi z različnimi vrednostmi hiperparametra _k_ in opazujte, kako število sosedov vpliva na točnost algoritma:

```python
acc_uni=[] #seznam za točnosti pri metodi uniformnega glasovanja
acc_dis=[] #seznam za točnosti za metodo glasovanja o razdalji
max_k=20 #maksimalna vrednost za k (vredno je nastavljati vrednosti do 150)

for k in range(max_k): 
    w = "uniform" # uniformno glasovanje
    clf = neighbors.KNeighborsClassifier(n_neighbors=k+1, weights=w) # ustvari k-NN klasifikator
    clf.fit(X, y) # nauči model
    pred = clf.predict(X) # napovej vrednosti za vse podatke
    correct = accuracy_score(pred,y,normalize=False) # izračunaj število pravilno klasificiranih primerov
    score = accuracy_score(pred,y,normalize=True) # izračunaj točnost
    print("Algoritem je določil",correct,"pravilno napovedanih primerov za",k+1,"-NN in",w,"metodo(točnost=","{:.2f}".format(score),").")
    acc_uni.append(score) # dodaj točnost v seznam, ki bo uporabljen za risanje
    
    w = "distance" # glasovanje o razdalji
    clf = neighbors.KNeighborsClassifier(n_neighbors=k+1, weights=w) # ustvari k-NN klasifikator
    clf.fit(X, y) # nauči model
    pred = clf.predict(X) # napovej vrednosti za vse podatke
    correct = accuracy_score(pred,y,normalize=False) # izračunaj število pravilno klasificiranih primerov
    score = accuracy_score(pred,y,normalize=True) # izračunaj točnost
    print("Model je določil",correct,"pravilno napovedanih primerov za",k+1,"-NN in",w,"metodo(točnost=","{:.2f}".format(score),").\n")
    acc_dis.append(score) # dodaj točnost v seznam, ki bo uporabljen za risanje
    
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1,max_k+1),acc_uni)
plt.xlabel("k")
plt.ylabel("Točnost")
plt.title("Uniform glasovanje")
plt.subplot(1, 2, 2)
plt.plot(range(1,max_k+1),acc_dis)
plt.xlabel("k")
plt.ylabel("Točnost")
plt.title("Glasovanje o razdalji")
plt.show() 
```

# 8.2. Linearna regresija

### 8.2.1. Opredelitev  

**Linearna regresija** je statistična metoda, ki pomaga napovedati razmerje med dvema spremenljivkama podatkovne točke: njeno značilnostjo in njeno vrednostjo. Cilj je poiskati premico (imenovano regresijska premica), ki se najbolje prilega podatkovnim točkam na razsevnem diagramu. To premico lahko nato uporabimo za napovedovanje prihodnjih vrednosti podobnih podatkovnih točk glede na njihovo značilnost. Uporablja se (seveda) za reševanje regresijskih problemov.  

Preprosto povedano, linearna regresija nam pomaga razumeti razmerje med dvema spremenljivkama in na podlagi tega razmerja napovedovati. To razmerje je podano z modelno hipotezo:  

$y_{pred}=\theta_{0}+\theta_{1}x$

Kot vidite, model ni nič drugega kot enačba za premico v koordinatnem sistemu $(x,y)$. Pri tem je $x$ atribut podatkovne točke, $y_{pred}$ pa napovedana vrednost podatkovne točke. Na primer, $x$ je lahko velikost hiše (v kvadratnih metrih), $y_{pred}$ pa je lahko napovedana cena te hiše. Z linearno regresijo dobimo model, ki poskuša najti povezavo med velikostjo in vrednostjo hiše. Računalnik to stori tako, da najde najboljše možne vrednosti **parametrov** $\theta_0$ in $\theta_1$.  

To nastavljanje parametrov je pravzaprav del algoritma, ki temelji na strojnem učenju. Računalnik potrebuje nabor učnih podatkov z veliko podatkovnimi točkami v obliki [velikost,cena]. Na podlagi teh podatkovnih točk algoritem poišče črto, ki jih najbolje opisuje, tako da izračuna, kako daleč je vsaka točka od črte (čim bližje, tem bolje). Črta, ki se čim bolj približa podatkovnim točkam, je optimalna črta (slika 3).

Te razdalje podatkovnih točk od črte predstavljajo razliko med dejansko vrednostjo podatkovne točke $y_{act}$ in vrednostjo podatkovne točke, ki jo je napovedal algoritem (črta) $y_{pred}$.  

### 8.2.2. Stroškovna funkcija in optimalni parametri  

Te stvari so nam morda vizualno jasne in lahko razumljive, vendar jih računalnik potrebuje prevedene v matematični jezik, da jih lahko interpretira in izvede. Tako pravimo, da kakovost modela predstavlja "srednja kvadratna napaka (MSE) vseh podatkovnih točk" $J$:  

$J(\theta_0,\theta_1)=\frac{1}{2m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)^2=\frac{1}{2m}\sum_{i=1}^{m}(\theta_0+\theta_1x^i-y_{act}^i)^2$

Tu je $m$ število podatkovnih točk v naboru podatkov za učenje. Manjša kot je ta skupna napaka, boljši je model. Algoritem torej dejansko poskuša najti parametra $\theta_0$ in $\theta_1$, ki minimizirata funkcijo $J$. Ti parametri bodo dali enačbo najboljše možne črte za dani niz podatkov.  

Ne da bi se spuščali v prevelike matematične podrobnosti, algoritem to doseže z usposabljanjem na naslednji način:  

$\text{initialize } \theta_0=\theta_1=0$  
$\text{za vnaprej določeno število korakov ali do konvergence ponovite naslednje:}$  
$\{$  
&emsp;$\theta_0=\theta_0-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)$  
&emsp;$\theta_1=\theta_1-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)x^i$  
$\}$

V tem primeru dosežena konvergenca pomeni, da se vrednost $J$ po določenem času preneha spreminjati. Tu je $\alpha$ hiperparameter, imenovan **stopnja učenja**, ki ima vrednost med 0 in 1 in določa, kako hitro bo potekalo učenje. Majhne vrednosti $\alpha$ bodo povzročile počasno, a varno učenje, medtem ko lahko prevelika vrednost destabilizira algoritem in ustvari ničvredne in neskončno velike vrednosti parametrov $\theta_0$ in $\theta_1$.  

Po končanem postopku dobimo **optimalne vrednosti** parametrov in najboljšo možno črto, ki opisuje podatke. Če nas zanima cena nove hiše, lahko v naš model vnesemo njeno velikost (njeno značilnost), preverimo, kakšna bo napovedana vrednost, in jo primerjamo z dejansko vrednostjo. Za ocenjevanje ponovno uporabimo MSE.  

### 8.2.3. Primer z enim samim atributom 

Ogledali si bomo primer uporabe linearne regresije za izgradnjo modela, ki napoveduje tedenski turistični dobiček (v tisočih EUR) majhnih mest na Balkanu v odvisnosti od števila njihovih prebivalcev. Prebivalstvo vsakega mesta je atribut $x$, dobiček vsakega mesta pa je vrednost $y$.  

Po uvozu, izpisu in vizualizaciji podatkov jih bomo razdelili na učne in testne podmnožice ter uporabili algoritem linearne regresije.

```python
# Vključi vse knjižnice:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = pd.read_csv('profit.txt', header=None, names=['Populacija', 'Profit']) # odpri datoteko
print ("Predogled podatkov:\n",data.head())
print ("Statistika podatkov:\n",data.describe())

data.plot(kind='scatter', x='Population', y='Profit', figsize=(8,4));   # nariši podatkovne točke
```

```python
X = data.iloc[:,0] # atributi se nahajajo v vseh vrsticah prvega stolpca
y = data.iloc[:,1] # vse ciljne vrednosti se nahajajo v vseh vrsticah drugega stolpca

# Razdeli podatke v učno in testno množico (razmerje 9:1): 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/10, random_state=0) 
# Izpiši dimenzije podatkov
print('Dataset size: ', data.shape)
print('X_train size: ', X_train.shape)
print('X_test size: ', X_test.shape)
print('y_train size: ', y_train.shape)
print('y_test size: ', y_test.shape)

# Poženi algoritem:
lr = LinearRegression()
lr.fit(X_train.values.reshape(-1,1),y_train) # nauči model
# Pridobi koeficiente: 
theta_0 = lr.intercept_ 
theta_1 = lr.coef_
print("Theta_0=","{:.2f}".format(theta_0))
print("Theta_1=","{:.2f}".format(theta_1[0]))
```

Po učenju algoritma na učni podmnožici vidimo, da je enačba za premico naslednja:  

$y_{pred}=-3.73+1.17x$  

Nato skozi to enačbo prenesemo testne podatke in izračunamo srednjo kvadratno napako kot kazalnik uspešnosti modela (čim manjša je, tem bolje).

```python
y_pred = lr.predict(X_test.values.reshape(-1,1)) # napovej
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred)) # ovrednoti
```

Na koncu narišimo črto in si oglejmo, kako zajema podatke. Dodali bomo tudi dodatno mesto in videli, kaj model pove o njem.  

```python
x = np.linspace(data.Population.min(), data.Population.max(), 100)
f = theta_0 + theta_1 * x # to je linearni model s theta parametri

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x, f, 'r') # nariši predvideno črto 
ax.scatter(data.Population, data.Profit) # nariši podatkovne točke
ax.set_xlabel('Populacija (v tisočih)')
ax.set_ylabel('Profit (v tisočih)')
ax.set_title('Napovedan Profit vs. Velikost Populacije')

prof = lr.predict([[20]]) # dodaj novo mesto s populacijo 20
print("Na primer, naš model je napovedal, da bi imelo mesto s populacijo 20.000 ljudi tedenski profit, ki je enak","{:.2f}".format(1000*prof[0]),"EUR.")
ax.scatter(20, prof, c='r'); # nariši mesto in njeno vrednost na istem grafu
```

### 8.2.4. Primer z več funkcijami  

V resničnih problemih so vrednosti redko odvisne od ene same funkcije. Dobiček mesta ni odvisen le od števila prebivalcev, temveč tudi od številnih drugih dejavnikov (velikost, lokacija, država, podnebje itd.). Podobno je cena hiše odvisna od več kot le od njene velikosti (tudi od števila sob, starosti, mesta, soseske itd.). Zato se linearna regresija zelo redko uporablja pri eni sami lastnosti, temveč pri več lastnostih.

Vendar pa se algoritem ne glede na število lastnosti sploh ne spremeni. Nabor podatkov ima lahko $n$ atributov in je zato n-dimenzionalen. Če je $n>3$, je ne bi mogli izrisati ali vizualizirati, vendar bo linearna regresija kljub temu delovala. Edina razlika je v tem, da model ne bo več podan z dvodimenzionalno premico, temveč z n-dimenzionalno hiperploščo.  

Model je zdaj podan z:  

$y_{pred}=\theta_{0}+\theta_{1}x_1+...+\theta_{n}x_n$  

To je tudi enačba hiper ravnine. To pomeni, da ima model na splošno $n+1$ parametrov.  

Tudi algoritem je podoben:  

$\text{initialize } \theta_0=\theta_1=...=\theta_n=0$  
$\text{za vnaprej določeno število korakov ali dokler ne dosežemo konvergence, ponovimo naslednje:}$  
$\{$  
&emsp;$\theta_0=\theta_0-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)$  
&emsp;$\theta_1=\theta_1-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)x_1^i$  
&emsp;...  
&emsp;$\theta_n=\theta_n-\alpha\frac{1}{m}\sum_{i=1}^{m}(y_{pred}^i-y_{act}^i)x_n^i$  
$\}$  

Tudi srednja kvadratna napaka je enaka. Linearna regresija se torej dobro obnese pri problemih z več značilnostmi, njeno delovanje pa lahko pokažemo v naslednjem primeru. Gre za primer z dvema značilnostma, da smo lahko še vedno vizualizirali podatke.  

Zdaj moramo sestaviti model, ki bo napovedal ceno hiše v EUR $y$, če poznamo velikost v kvadratnih metrih (značilnost $x_1$) in število spalnic (značilnost $x_2$). Opazite, da uporabljamo skoraj enako kodo kot prej.  

Vendar je opazna razlika: podatke moramo najprej **standardizirati**. Usklajevanje pomeni, da moramo od vsake podatkovne točke odšteti povprečje podatkov in nato rezultat deliti s standardnim odklonom podatkov. Tako dobimo nove podatke, ki vsebujejo enake informacije kot prej, vendar imajo zdaj povprečje 0 in standardni odklon 1. To je treba storiti, kadar delamo z več značilnostmi, saj so te značilnosti na splošno različno razporejene v prostoru stanja. Na primer, značilnost, ki označuje velikost hiše, se lahko giblje od 800 do 4500 kvadratnih metrov, značilnost, ki označuje število sob, pa se giblje le od 1 do 5. Da bi se izognili tej neenakomerni porazdelitvi, ki lahko povzroči napake in upočasni izvajanje algoritma, moramo standardizirati.

```python
# Vključi vse knjižnice:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = pd.read_csv('prices.txt', header=None, names=['Size', 'Bedrooms', 'Price']) # open data file
print ("Predogled podatkov:\n",data.head())
print ("Statistika podatkov:\n",data.describe())

# Standardiziraj podatke
data_s = (data - data.mean() ) / data.std()
print ("Standardizirani podatki:")
print (data_s.head())
print (data_s.describe())

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data_s.Size, data_s.Bedrooms, data_s.Price)  # nariši vse podatkovne točke
```

```python
X = data_s.iloc[:,:2] # vsi atributi so vsebovani v prvih dveh stolpcih
y = data_s.iloc[:,2] # ciljne vrednosti so vsebovane v tretje stolpcu

# Razdeli podatke na učno in testno množico (razmerje 9:1): 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/10, random_state=0) 
# Izpiši dimenzije vseh podatkov in množic 
print('Dataset size: ', data.shape)
print('X_train size: ', X_train.shape)
print('X_test size: ', X_test.shape)
print('y_train size: ', y_train.shape)
print('y_test size: ', y_test.shape)

# Poženi algoritem
lr = LinearRegression()
lr.fit(X_train,y_train) # nauči model
# Pripravi koeficiente:
theta_0 = lr.intercept_ 
theta_1 = lr.coef_[0]
theta_2 = lr.coef_[1]
print("Theta_0=","{:.2f}".format(theta_0))
print("Theta_1=","{:.2f}".format(theta_1))
print("Theta_2=","{:.2f}".format(theta_2))
```

Po usposabljanju algoritma na učni podmnožici vidimo, da je enačba za dvodimenzionalno ravnino naslednja:  

$y_{pred}=-0.07+0.81x_1-0.03x_2$  

Nato skozi to enačbo prenesemo testne podatke in izračunamo srednjo kvadratno napako kot kazalnik uspešnosti modela (čim manjša je, tem bolje).

```python
y_pred = lr.predict(X_test) # napovej
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred)) # ovrednoti
```

Na koncu narišimo ravnino in si oglejmo, kako zajema podatke. Dodali bomo tudi dodatno hišo in videli, kaj model pove o njej.

```python
x12 = np.linspace(data_s.Size.min(), data_s.Size.max(), 100)
x22 = np.linspace(data_s.Bedrooms.min(), data_s.Bedrooms.max(), 100)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data_s.Size, data_s.Bedrooms, data_s.Price, c='c')  # narišimo podatkovne točke
X12,X22 = np.meshgrid(x12,x22)

f = theta_0 + theta_1 * X12 + theta_2 * X22 # linearni model z theta parametri

ax.plot_surface(X12, X22, f) # nariši predvideno črto
ax.set_xlabel('Velikost - standardizirano')
ax.set_ylabel('Sobe - standardizirano')
ax.set_zlabel('Cena [EUR] - standardizirano')
ax.set_title('Napovedana cena vs. Velikost Hiše')

new_house = data.iloc[-1,:] # vzami novo hiši (zadnji primer iz podatkov, npr. 3 spalnice)
new_house = (new_house - data.mean() ) / data.std() # standardiziraj
new_X = new_house[:2] # take only the features
price = lr.predict(new_X.values.reshape(1,-1)) # napovej novo vrednost 
ax.scatter(new_X[0], new_X[1], price, c='r'); # nariši hišo in njeno ceno na isti sliki
```

Točka je točno na ravnini, vendar se v bližini rahlo razlikuje od dejanske vrednosti.  

Spet je očitno, da ravnina ne pokriva vseh točk popolnoma, kar je nemogoče, saj niso linearne. Vendar pa uspešno zajame trend podatkov, ki je blizu linearnemu. To je največ, kar lahko zahtevamo od linearne regresije, ko jo uporabljamo za reševanje resničnih, nelinearnih problemov.

## 8.3. Logistična regresija  

### 8.3.1. Opredelitev  

**Logistična regresija** je statistična metoda, ki se uporablja pri strojnem učenju za napovedovanje binarnega rezultata (tj. dveh možnih rezultatov, na primer da/ne ali res/ne) na podlagi ene ali več vhodnih spremenljivk. Kot smo že obravnavali, gre pravzaprav za postopek binarne klasifikacije.  

Predstavljajte si, da poskušate napovedati, ali bo študent opravil izpit ali ne, na podlagi števila ur, ki jih je študiral. Te informacije lahko narišete na graf, pri čemer je na osi X število ur študija, na osi Y pa rezultat uspešno/neuspešno opravljen izpit.  

<figure>
<img src="https://github.com/VALENCEML/eBOOK/raw/main/SL/logreg.jpeg" style="height: 400px;"/> 
<figcaption align ="center"> Slika 4 (via Wikimedia Commons) </figcaption>
</figure>  

V zgornjem grafu pike predstavljajo posamezne študente ter njihove ure študija in rezultate izpitov. Opazite, da se s povečevanjem števila študijskih ur povečuje tudi verjetnost opravljanja izpita. Tu nastopi logistična regresija - pomaga nam poiskati najprimernejšo premico, ki čim bolje loči oba izida. Opazite, da je črta zdaj ukrivljena, za razliko od ravne črte, ki smo jo imeli pri linearni regresiji.  

Linijo lahko matematično predstavimo z enačbo, ki sprejme vhodno spremenljivko (v tem primeru število ur študija) in izpiše verjetnost opravljanja izpita. Enačba za premico v zgornjem grafu je na primer lahko videti takole:  

$y_{pred}=\frac{1}{1+e^{-(\theta_0+\theta_1x)}}$  

To je model ali hipoteza. Tu je $x$ ura študija (značilnost), $\theta_0$ in $\theta_1$ pa sta spet parametra, ki ju mora algoritem izračunati za najboljšo možno linijo. Nazadnje je $y_{pred}$ vrednost, ki jo bo ta model napovedal - opazite, da se giblje med 0 (če je eksponent $e$ v oklepaju zelo negativen, je imenovalec pozitivno neskončen) in 1 (če je eksponent $e$ v oklepaju zelo pozitiven, je imenovalec 1). Zato model v logistični regresiji izpiše vrednost verjetnosti med 0 in 1 (0 % in 100 %) o tem, kako verjeten je izid problema. Ker pa gre za binarno klasifikacijo, pravimo, da je sredina tega razpona 0,5 (50 %) prag pri odločanju:  
* če $y_{pred}>=0,5$, potem model napoveduje 1 ali TRUE;  
* če $y_{pred}<0,5$, potem model napoveduje 0 ali PRAVDA.  

Ta pristop se imenuje logistična regresija, čeprav se uporablja samo za probleme klasifikacije. Razlog za to je, da je model spet regresijski model (izpiše stalno vrednost $y_{pred}$), kot je razvidno iz slike 4. Vendar je uvedba praga, ki smo ga opredelili zgoraj, tista, zaradi katere ta pristop postane klasifikacijski pristop (izhodni rezultati so 0 ali 1). Ime "logistična" izhaja iz vrste funkcije, uporabljene v modelu, ki se imenuje logistična funkcija.  

Podobno kot pri linearni regresiji se lahko tudi logistična regresija uporablja pri problemih z več značilnostmi, kar je dobro, saj ima večina resničnih problemov veliko več kot en atribut. V primeru $n$ lastnosti (n-razsežni problem) je model naslednji:  

$y_{pred}=\frac{1}{1+e^{-(\theta_0+\theta_1x_1+...+\theta_nx_n)}}$  

Ne da bi se spuščali v matematične podrobnosti, računalnik optimizira parametre $\theta$ na podoben način, kot je prikazano pri linearni regresiji. 

Na koncu je način ocenjevanja uspešnosti modela logistične regresije podoben načinu, ki je bil prej predstavljen pri k-NN. Preštejemo vse pravilno razvrščene primere iz podmnožice testnih podatkov in to število delimo s skupnim številom podatkovnih točk v testni podmnožici, s čimer dobimo metriko natančnosti.  

$\text{TOČNOST}=\frac{\text{ŠTEVILO PRAVILNO NAPOVEDANIH PODATKOVNIH TOČK}}{\text{ŠTEVILO VSEH PODATKOVNIH TOČK}}$

### 8.3.2. Primer razvrščanja z več spremenljivkami  

Oglejmo si primer, v katerem imamo podatke o 100 študentih, ki so se prijavili na tehnično univerzo. Poznamo njihove rezultate dveh maturitetnih izpitov iz matematike in fizike (to sta atributa $x_1$ in $x_2$) in vemo, ali so bili sprejeti (1) ali niso bili sprejeti (0) na univerzo (to je razred). Usposobimo logistični regresijski klasifikator, ki zgradi model in lahko na podlagi rezultatov maturitetnih izpitov predvidi, ali se lahko prihajajoče generacije študentov vpišejo.

```python
# Vključimo vse knjižnice in podatke
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix

data = pd.read_csv('https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/exams.txt', header=None, names=['MathScore', 'PhysicsScore', 'Accepted?'])
print("Vsi podatki:\n",data)
X = data.drop(columns=['Accepted?']) # x is all columns but 'Accepted?'
y = data['Accepted?'] # y is the 'Accepted?' column 
print("X:\n",X)
print("y:\n",y)

# Plot the data
positive = data[data['Accepted?'].isin([1])]  
negative = data[data['Accepted?'].isin([0])]
fig, ax = plt.subplots(figsize=(8,4))  
ax.scatter(positive['MathScore'], positive['PhysicsScore'], s=50, c='b', marker='o', label='Accepted')  
ax.scatter(negative['MathScore'], negative['PhysicsScore'], s=50, c='r', marker='x', label='Not Accepted')  
ax.legend()  
ax.set_xlabel('Ocena pri matematiki')  
ax.set_ylabel('Ocena pri fiziki');
```

Vidimo lahko, da so podatki precej pričakovani - tisti z višjimi ocenami so na splošno sprejeti na univerzo. Toda poglejmo, kje točno je ločnica. Sedaj bomo usposobili logistični regresijski model.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/10, random_state=0) # get train and test subsets
model = LogisticRegression() # ustvari klasifikator
model.fit(X_train, y_train) # nauči model
# Izpiši theta koeficiente: 
print("Theta_0 = ","{:.2f}".format(model.intercept_[0]))
print("Theta_1 = ","{:.2f}".format(model.coef_[0,0]))
print("Theta_2 = ","{:.2f}".format(model.coef_[0,1]))
```

Po usposabljanju algoritma na učni podmnožici vidimo, da je model:  

$y_{pred}=\frac{1}{1+e^{-(-24.14+0.19x_1+0.20x_2)}}=\frac{1}{1+e^{(24.14-0.19x_1-0.20x_2)}}$ 

Nato s to enačbo preverimo testne podatke, da dobimo napovedi, in izračunamo natančnost v primerjavi z dejanskimi vrednostmi.

```python
y_pred = model.predict(X_test)
acc = metrics.accuracy_score(y_test,y_pred)
print("Accuracy=","{:.2f}".format(acc))
```

Te rezultate predstavimo tudi na drugačen način. Natančnost daje le eno vrednost za količinsko opredelitev modela, če pa uporabimo **matriko zmede**, lahko dobimo več informacij. Matrika zmede ima 2 vrstici in 2 stolpca. Vrstice predstavljajo prave vrednosti razredov (0 v vrstici 0 in 1 v vrstici 1). Stolpci predstavljajo predvidene vrednosti razredov (0 v vrstici 0 in 1 v vrstici 1).  
* Številka na mestu (0,0) predstavlja **resnično negativno vrednost TN**. To je število podatkovnih točk, ki so bile iz razreda 0 in so bile pravilno napovedane kot razred 0.
* Število na mestu (0,1) predstavlja **nepravilno pozitivne rezultate FP**. To je število podatkovnih točk iz razreda 0, ki so bile napačno napovedane kot razred 1.
* Število na mestu (1,0) predstavlja **nepravilno negativne podatke FN**. To je število podatkovnih točk iz razreda 1, ki so bile napačno napovedane kot razred 0.
* Številka na mestu (1,1) predstavlja **pravilno pozitivne rezultate TP**. To je število podatkovnih točk iz razreda 1, ki so bile pravilno napovedane kot razred 1.

```python
plot_confusion_matrix(model,X_test,y_test,cmap='Reds');
```

Kot vidimo, je bilo v testni podmnožici skupaj 10 podatkovnih točk, 5 pozitivnih in 5 negativnih. Vseh 5 pozitivnih podatkov je bilo pravilno razvrščenih (glej desni stolpec), pri razvrščanju 5 negativnih podatkov pa je prišlo do ene napake (glej levi stolpec). To potrjuje predhodno doseženo 90-odstotno natančnost. Matrika zmede je uporabno orodje pri vizualizaciji natančnosti in je veliko bolj informativna od same ocene natančnosti. Pogosto se uporablja pri vseh klasifikacijskih algoritmih.  

#### 8.3.2.1. Ločitvena črta  

Na koncu je treba omeniti še eno pomembno stvar. Čeprav model logistične regresije ne daje več ravne črte, se v njem skriva ravna črta (ali v splošnem primeru hiper ravnina). Eksponent $e$ znotraj logistične funkcije je pravzaprav enačba za premico ali ravnino, ki ločuje podatke v prostoru funkcij. Da bi to pokazali, narišimo v funkcijskem prostoru premico, ki jo tvorijo parametri, ki smo jih dobili.  

$x_2 = -\frac{\theta_0}{\theta_2}-\frac{\theta_1}{\theta_2}x_1=\frac{24.14}{0.20}-\frac{0.19}{0.20}x_1$

```python
x = np.linspace(data.MathScore.min(), data.MathScore.max(), 100)
f = -model.intercept_[0]/model.coef_[0,1] - model.coef_[0,0]*x/model.coef_[0,1]
fig, ax = plt.subplots(figsize=(8,4))  
ax.scatter(positive['MathScore'], positive['PhysicsScore'], s=50, c='b', marker='o', label='Accepted')  
ax.scatter(negative['MathScore'], negative['PhysicsScore'], s=50, c='r', marker='x', label='Not Accepted')  
ax.legend()  
ax.set_xlabel('Ocena pri matematiki')  
ax.set_ylabel('Ocena pri fiziki');
ax.plot(x, f, 'g', label='Decision Boundary');
```

Očitno je, da se nekatere podatkovne točke nahajajo na napačni strani premice, vendar je to pričakovano, saj podatkov ni mogoče linearno ločiti. Ta ločitvena črta (imenovana tudi **mejna črta odločanja**) bo vedno ravna, zato bo algoritem delal napake. Če pa je meja skoraj linearna z nekaterimi izjemami, bo logistična regresija delovala odlično (opazite podobnost z linearno regresijo v tem smislu).  

### 8.3.3. Klasifikacija več razredov  

Logistično regresijo in tudi druge metode klasifikacije je mogoče uporabiti tudi za razvrščanje podatkov, ki pripadajo več kot dvema razredoma. To storimo z uporabo tako imenovane metode "eden proti vsem" (slika 5). Ideja je, da če obstaja $k$ možnih razredov, se usposobi $k$ različnih klasifikatorjev - vsak bo enega od razredov obravnaval kot pozitivnega (1), vse druge pa kot negativne (0). Pri testiranju novih podatkov se nato podatkovne točke preženejo skozi vseh $k$ klasifikatorjev, pri čemer vsak od njih izpiše verjetnost, kako verjetno je, da točka pripada njegovemu razredu. Klasifikator, ki poda najvišjo verjetnost, "zmaga" in podatkovna točka je razvrščena v svoj razred.

## 8.4. Decision trees 

### 8.4.1. Opredelitev  

Drevo odločanja** je vrsta algoritma strojnega učenja, ki se lahko uporablja za reševanje problemov z napovedovanjem na podlagi podatkov. Deluje tako, da zapleten problem razdeli na manjše, lažje obvladljive dele in nato sprejema odločitve na podlagi podatkov, ki so na voljo na vsakem koraku. Ne zgradi eksplicitnega modela za vse podatke vnaprej in v tem smislu se razlikuje od logistične in linearne regresije. Uporablja se lahko tako za klasifikacijo kot za regresijo, vendar jo je lažje razumeti in predstaviti v okviru klasifikacije, zato bomo to storili tukaj.  

Predstavljajte si, da želite napovedati vrsto sadja, ki ga imate, na podlagi njegove barve in oblike. Začnete lahko z vrsto vprašanj, kot so:

Ali je sadje mehko?
* a. Da:
  * i. Ali je sadež rdeč?
    * 1. Da: sadež je jagoda.
    * 2. Ne: sadež je banana.
* b. Ne:
  * i. Ali je sadež rdeč?
    * 1. Da: sadež je granatno jabolko.
    * 2. Ne: sadež je ananas.  
    
Vsak korak v postopku pomaga zožiti možne izbire, dokler ni mogoče narediti končne napovedi. To je podobno delovanju odločitvenega drevesa pri strojnem učenju, kjer vsak korak predstavlja **vozlišče** v drevesu, odločitve, sprejete v vsakem vozlišču, pa vodijo v različne **vejitve**. Končna napoved se pripravi v **listnem vozlišču**. Drevo odločanja je torej mogoče prikazati kot drevesu podoben diagram, kjer veje predstavljajo različne odločitve, sprejete v vsakem vozlišču, listi pa končne napovedi.

V tem primeru je prva odločitev, ali je sadež rumen ali ne. Če je rumen, algoritem preveri, ali je dolg in vitek ali ne. Kot vidite, je krmarjenje po odločitvenem drevesu zelo preprosto, saj vse, kar počnemo, je, da odgovarjamo na vprašanja v izbranih vejah glede na naše odgovore, dokler ne pridemo do lista. Natanko tako je videti faza testiranja, ko se podatkovne točke iz testne podmnožice vnesejo v drevo in ga "prečkajo" glede na vrednosti svojih lastnosti, dokler ne pridejo do lista, ki vsebuje njihov razred/vrednost. Vendar je težavno vprašanje: kako je drevo zgrajeno z uporabo podatkov za učenje, tj. kako so izbrana vprašanja v vozliščih?

### 8.4.2. Onesnaženost in gradnja drevesa  

Odločitvena drevesa uporabljajo metriko, imenovano **gini index**, za odločanje o tem, katere podatke razdeliti, tj. po katerih lastnostih in vrednostih razvejati vozlišča. Gini index je merilo, kako "mešana" je skupina stvari. Predstavljajte si, da imate skledo bombonov M&M in jih želite razdeliti v dve ločeni skledi glede na barvo. Cilj je imeti eno skledo z vsemi modrimi bonboni in drugo skledo z vsemi rdečimi bonboni. Bolj kot so bonboni pomešani, višja bo Ginijeva ocena nečistoče. Če sta posodi popolnoma razdeljeni po barvah, bo Ginijeva ocena nečistosti 0, kar pomeni, da bonboni sploh niso pomešani. Pri strojnem učenju uporabljamo Ginijevo nečistost za ocenjevanje, kako dobro je mogoče niz predmetov razdeliti v različne skupine na podlagi določenih značilnosti. Cilj je imeti nizko oceno Ginijeve nečistosti, kar pomeni, da so predmeti dobro razdeljeni v skupine s podobnimi značilnostmi.  

Pri usposabljanju drevesa računalnik poskuša vsako posamezno značilnost uporabiti kot korensko vozlišče. Nato jih razveji in po razdelitvi izračuna stopnjo nečistoče. Značilnost, ki daje najnižjo ginijevo oceno nečistoče, se izbere kot koren in se razveji. Nato se postopek ponovno ponovi v vseh vozliščih, v vsaki posamezni globini in za vse lastnosti. To se nadaljuje, dokler ne dosežemo vnaprej določene največje globine ali dokler nimamo več vozlišč za razvejitev (tj. dokler ne dosežemo vseh listnih vozlišč). Pridobitev listnega vozlišča pomeni, da imamo ginijevo nečistost 0, saj so bili vsi podatki v tem vozlišču jasno in pravilno razvrščeni.  

Ta pristop se vam bo morda zdel dolgotrajen in naporen - preizkušanje vseh lastnosti v vsakem vozlišču se sliši utrujajoče! Vendar ne pozabite, da globlje ko gremo v drevo, manj podatkov imamo na voljo, saj vsako vozlišče analizira samo podatke, ki izpolnjujejo zahteve vej, ki vodijo do njega. Tako stvari potekajo hitreje in računalnik si lahko privošči, da vedno znova preizkuša različne funkcije.

Preizkusimo več različnih odločitvenih dreves z naborom podatkov o šarenici, da ponazorimo vse, kar smo povedali zgoraj. Tako kot prej bomo uvozili podatke, jih razdelili na učne in testne podmnožice, nato pa usposobili odločitvena drevesa in preizkusili njihovo natančnost. Preizkusili bomo tri različna odločitvena drevesa in primerjali njihove natančnosti: eno z največjo globino 2, eno z največjo globino 5 in eno brez določene največje globine (tj. razveji se, dokler ne doseže vseh listnih vozlišč).

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# naložimo podatkovno zbirko iris
iris = load_iris()

# ustvarimo podatkovni okvir za atribute in ciljno vrednost
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
print(df)

# razdelimo podatke učno in testno množico
X_train,X_test,y_train,y_test = train_test_split(df[iris['feature_names']],df['target'],test_size=0.25,random_state=0)

# naučimo model z maksimalno globino 2:
clf1 = DecisionTreeClassifier(max_depth=2, random_state=0)
clf1.fit(X_train, y_train)
# Ovrednotimo model na testni množici
y_pred1 = clf1.predict(X_test)
print("Accuracy with max_depth=2:", accuracy_score(y_test, y_pred1))

# naučimo model z maksimalno globino 5:
clf2 = DecisionTreeClassifier(max_depth=5, random_state=0)
clf2.fit(X_train, y_train)
# Ovrednotimo model na testni množici
y_pred2 = clf2.predict(X_test)
print("Accuracy with max_depth=5:", accuracy_score(y_test, y_pred2))

# naučimo model brez omejitve globine
clf3 = DecisionTreeClassifier(max_depth=None, random_state=0)
clf3.fit(X_train, y_train)
# Ovrednotimo model na testni množici
y_pred3 = clf3.predict(X_test)
print("Accuracy with max_depth=None:", accuracy_score(y_test, y_pred3)) 
```

Vidimo, da je najboljša natančnost dosežena pri največji globini 5 (ali morda celo prej), vendar največja globina 2 ni dovolj. Za ponazoritev tega narišimo odločitveno drevo z največjo globino in poglejmo, kako globoko gre.

```python
import matplotlib.pyplot as plt
# nariši drevo brez omejitve globine (clf3):
plt.figure(figsize=(20, 20))
tree.plot_tree(clf3, filled=True, feature_names=df.columns)
plt.show()
# levo -> drži; desno >- ne drži
```

Drevo najprej razdeli koren na podlagi širine cvetnega lista in določi odločitveni prag 0,8 cm. Cvetovi z manjšo širino cvetnega lista se takoj razvrstijo v razred 1 iris, ker naletimo na podmnožico podatkov za učenje, ki ne vsebuje drugih cvetov (ima gini nečistočo 0). Zato je to list. Preostali podatki se nato razdelijo glede na značilnost dolžina cvetnega lista, pri čemer je prag 4,95 cm, in tako naprej. Kot lahko vidimo, ima drevo globino 4, preden doseže vsa listna vozlišča z ginijevo nečistočo 0, torej je to največja globina.  

To je ena od velikih prednosti odločitvenih dreves - jasno lahko vidimo in razumemo, zakaj razvrščajo podatke na določen način. Vse, kar moramo storiti, je narisati drevo ter slediti vozliščem in vejam za vsak vzorec. Zato se odločitvena drevesa imenujejo "model bele škatle" (kot v "prozorni škatli").  

Vidimo lahko tudi, da se ginijeva nečistoča zmanjšuje, ko se spuščamo globlje in stvari postajajo jasnejše - začnemo z nečistočo 0,665 pri korenu in končamo z 0 nečistočami na vseh listih. Opazite tudi, kako začnemo s skupno 112 vzorci (učni podatki so 3/4 prvotnega nabora podatkov s 150 cvetovi), vendar se število vzorcev zmanjšuje, ko gremo globlje v drevo.  

Nato si oglejmo matriko zmede za testno podmnožico na tem drevesu in tudi na drevesu z največjo globino 2, ki je imelo nižjo oceno natančnosti.

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf3,X_test,y_test,cmap='Reds');
plot_confusion_matrix(clf1,X_test,y_test,cmap='Reds');
```

Od 38 vzorcev, ki so bili uporabljeni za testiranje drevesa največje globine, je bil le en vzorec napačno razvrščen (napovedan razred 2, dejansko pa je bil razred 1). Drevo z največjo globino 2 je naredilo še 3 napake.   

Upoštevajte, da povečanje največje globine drevesa na splošno vodi do boljšega delovanja na podatkih za učenje, vendar lahko povzroči **pretirano prilagajanje** (tj. model je preveč zapleten in se ne posplošuje dobro na nove podatke). Dobra praksa je, da prilagodite največjo globino, da najdete najboljše ravnovesje med natančnostjo in posploševanjem.

### 8.4.3. Večji primer    

Poskusimo s primerom z večjo zbirko podatkov z več atributi in primeri. Vsebuje zdravstvene in fiziološke informacije o več bolnikih (značilnosti) ter informacijo o tem, ali je bila pri njih diagnosticirana sladkorna bolezen ali ne (razred 1 ali 0). Če usposobimo dober klasifikacijski model, ga lahko uporabimo za napovedovanje ali vsaj pomoč pri diagnosticiranju sladkorne bolezni pri novih bolnikih v prihodnosti.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# naloži podatkovno zbirko:
data = pd.read_csv("https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/diabetes.csv")

# razdeli podatke na atribute in ciljno vrednost:
X = data.drop("Outcome", axis=1)
y = data["Outcome"]

# razdeli podatke na učno in testno množico
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# nauči model
dtree = DecisionTreeClassifier(max_depth=None, random_state=0)
dtree.fit(X_train, y_train)

# napovej vrednosti na testni množici
y_pred = dtree.predict(X_test)

# izračunaj točnost:
accuracy = accuracy_score(y_test, y_pred)
print("Točnost: ","{:.2f}".format(accuracy))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(20, 20))
tree.plot_tree(dtree, filled=True, feature_names=X.columns)
plt.show()
# leva -> drži; desna >- ne drži
```

Drevo je zdaj veliko večje in je doseglo največjo globino 16. Oglejmo si, kako se spreminja natančnost testiranja pri različnih globinah.

```python
acc_train = [] # seznam za točnosti pri učnih primerih
acc_test = [] # seznam za točnosti pri testnih primerih

for i in range(16): # poskusi globino od 1-16
    dtree = DecisionTreeClassifier(max_depth=i+1, random_state=0) # ustvari drevo z določeno globino
    dtree.fit(X_train, y_train) # nauči model
    y_pred_train = dtree.predict(X_train) # testiraj na učnih podatkih
    y_pred_test = dtree.predict(X_test) # testiraj na testnih podatkih
    accuracy_train = accuracy_score(y_train, y_pred_train) # izračunaj točnost na učnih podatkih
    accuracy_test = accuracy_score(y_test, y_pred_test) # izračunaj točnost na testnih podatkih
    acc_train.append(accuracy_train) 
    acc_test.append(accuracy_test) 

# nariši točnosti za določeno globino:
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1,17),acc_train)
plt.xlabel("Maksimalna globina")
plt.ylabel("Točnost")
plt.title("Tree training accuracy depending on maximum depth")
plt.title("Točnost na učnih podatkih glede na maksimalno globino")
plt.subplot(1, 2, 2)
plt.plot(range(1,17),acc_test)
plt.xlabel("Maksimalna globina")
plt.ylabel("Točnost")
plt.title("Točnost na testnih podatkih glede na maksimalno globino")
plt.show() 
```

Kot je bilo pričakovati, če drevo preizkusimo na podatkih za usposabljanje (ki jih je že videlo med usposabljanjem), se natančnost izboljšuje z globino in doseže celo 100 %.  

Zanimivo pa je videti, da pri testiranju na nevidenih podatkih najboljša točnost ne doseže najglobljega drevesa. Slednja se pojavi pri globini 10. Razlog za to je, da globlja drevesa običajno preveč prilegajo podatkom, to pomeni, da "preveč razmišljajo" pri iskanju realnega razmerja med značilnostmi in vrednostjo/razredom. V resničnem svetu bi bil ekvivalent temu učenec, ki se nauči vprašanja in odgovore na pamet; če se mu zastavijo ta vprašanja, bo odlično poznal odgovore. Če pa se mu zastavi katero koli drugo vprašanje, povezano s snovjo, vendar drugačno od tistega, ki se ga je naučil, bo imel težave, ker se je vprašanja in odgovore naučil, ne da bi jih razumel.  

Kot smo že povedali, želimo dobro ravnovesje med natančnostjo in posploševanjem, zato moramo preveriti natančnost za različne vrednosti hiperparametrov (ki je v tem primeru največja globina drevesa), preden se odločimo za najboljšo.
